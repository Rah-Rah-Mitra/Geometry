# Mathematical Foundations of Geometric Deep Learning - Visualization-First Notebook Course

This standalone course follows the structure of *Mathematical Foundations of Geometric Deep Learning* using original prose, generated diagrams, executable checks, and book-local artifacts. The PDF is used only for source orientation.

## How To Use This Course

Each chapter folder contains a local `00-index.ipynb` and one canonical teaching notebook. Generated artifacts live under `artifacts/chapter-XX/` and are regenerated by the canonical notebooks.

## Course Map

- [Chapter 01: Algebraic Structures and Mathematics before Numbers](chapter-01-algebraic-structures/00-index.ipynb) - [canonical](chapter-01-algebraic-structures/01-algebraic-structures.ipynb); PDF pp. 4-19; sets, maps, power sets, Cartesian products, finite groups, Cayley graphs, actions, invariance/equivariance, tensor notation, einsum checks
- [Chapter 02: Geometric and Analytical Structures](chapter-02-geometric-analytical-structures/00-index.ipynb) - [canonical](chapter-02-geometric-analytical-structures/02-geometric-analytical-structures.ipynb); PDF pp. 20-25; norm balls, induced metrics, pseudometric/quasi-metric examples, Hausdorff distance, proximity graphs, projections, Gram-Schmidt, attention scaling
- [Chapter 03: Vector Calculus](chapter-03-vector-calculus/00-index.ipynb) - [canonical](chapter-03-vector-calculus/03-vector-calculus.ipynb); PDF pp. 26-36; Lipschitz/smoothness gallery, scalar/vector fields, gradients, Jacobians, integrals, divergence/flux, Laplacian, loss landscapes, backprop graph
- [Chapter 04: Topological Foundations and Differential Geometry](chapter-04-topology-differential-geometry/00-index.ipynb) - [canonical](chapter-04-topology-differential-geometry/04-topology-differential-geometry.ipynb); PDF pp. 37-50; open sets, continuity via preimages, homeomorphism/homotopy, Euler characteristic, geodesics, charts, tangent spaces, exp/log maps, gauges, manifold hypothesis
- [Chapter 05: Functional Analysis](chapter-05-functional-analysis/00-index.ipynb) - [canonical](chapter-05-functional-analysis/05-functional-analysis.ipynb); PDF pp. 51-54; Cauchy sequences, completeness gaps, Banach/Hilbert geometry, function bases, orthonormal reconstruction, operators, adjoints, functionals
- [Chapter 06: Spectral Theory](chapter-06-spectral-theory/00-index.ipynb) - [canonical](chapter-06-spectral-theory/06-spectral-theory.ipynb); PDF pp. 55-64; eigen-directions, self-adjoint spectra, Laplacian eigenfunctions, Fourier Gram matrices, SVD, Parseval, heat flow, graph Laplacian bridge
- [Chapter 07: Graph Theory](chapter-07-graph-theory/00-index.ipynb) - [canonical](chapter-07-graph-theory/07-graph-theory.ipynb); PDF pp. 65-77; graph anatomy, adjacency/degree matrices, graph distances, graph families, geometric graphs, homophily, permutations, homomorphisms, Laplacians, graph Fourier modes, message passing
