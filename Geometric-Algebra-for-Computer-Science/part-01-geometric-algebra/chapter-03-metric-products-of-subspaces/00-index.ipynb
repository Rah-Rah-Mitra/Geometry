## Chapter Coverage

- metrics, norms, and angles for blades
- scalar product of same-grade subspaces
- left and right contractions as metric grade-lowering products
- orthogonality, inverses, direct representation, and dual representation
- orthogonal projection of vectors and subspaces
- the 3-D cross product as a dual construction
- reciprocal frames and color-space conversion


## Study Strategy

Run the notebook once from top to bottom, then change the vectors in the metric, projection, and color-space cells. The best check is not whether a number stays the same, but whether the stated invariant still holds.
