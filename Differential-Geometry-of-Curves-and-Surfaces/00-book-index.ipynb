# Differential Geometry of Curves and Surfaces - Standalone Notebook Course

This course is an original executable notebook edition of Manfredo P. do Carmo's *Differential Geometry of Curves and Surfaces*, Revised & Updated Second Edition. It uses the local PDF only for source orientation and teaches the material through original prose, generated diagrams, computational experiments, and sanity checks.

## Source Notes

- The local PDF is used only for source orientation.
- Body printed pages map to PDF pages by pdf_page = printed_page + 16.
- Back matter is inventoried but not converted into copied hints or exercise text.

## Course Units

- [Chapter 01: Curves](chapter-01-curves/00-index.ipynb) - [canonical notebook](chapter-01-curves/01-curves.ipynb); printed pp. 1-52; PDF pp. 17-68; parametrized curves, regularity, arc length, Frenet frames, curvature, torsion, and global plane-curve behavior.
- [Chapter 02: Regular Surfaces](chapter-02-regular-surfaces/00-index.ipynb) - [canonical notebook](chapter-02-regular-surfaces/02-regular-surfaces.ipynb); printed pp. 53-135; PDF pp. 69-151; surface patches, tangent planes, differentiable maps, first fundamental form, area, and orientation.
- [Chapter 03: The Geometry of the Gauss Map](chapter-03-the-geometry-of-the-gauss-map/00-index.ipynb) - [canonical notebook](chapter-03-the-geometry-of-the-gauss-map/03-the-geometry-of-the-gauss-map.ipynb); printed pp. 136-219; PDF pp. 152-235; Gauss map, shape operator, curvatures, vector fields, ruled surfaces, and minimal surfaces.
- [Chapter 04: The Intrinsic Geometry of Surfaces](chapter-04-the-intrinsic-geometry-of-surfaces/00-index.ipynb) - [canonical notebook](chapter-04-the-intrinsic-geometry-of-surfaces/04-the-intrinsic-geometry-of-surfaces.ipynb); printed pp. 220-320; PDF pp. 236-336; isometries, conformal maps, compatibility equations, geodesics, parallel transport, Gauss-Bonnet, and exponential maps.
- [Chapter 05: Global Differential Geometry](chapter-05-global-differential-geometry/00-index.ipynb) - [canonical notebook](chapter-05-global-differential-geometry/05-global-differential-geometry.ipynb); printed pp. 321-474; PDF pp. 337-490; rigidity, completeness, variations, Jacobi fields, coverings, total curvature, flat surfaces, abstract surfaces, and Hilbert obstruction.
