# Part 2: Working and Playing with Geometry

Quadrilateral sets, conics, projective d-space, tensor diagrams, configurations, and theorem proving.

## Chapters

- Chapter 8: [Quadrilateral Sets and Liftings](chapter-08-quadrilateral-sets-and-liftings/00-index.ipynb) - Points on a line, quadrilateral sets, symmetry, von Staudt links, slope conditions, and involutions.
- Chapter 9: [Conics and Their Duals](chapter-09-conics-and-their-duals/00-index.ipynb) - Conic equations, polars, tangents, dual quadratic forms, transformations, degenerate conics, and primal-dual pairs.
- Chapter 10: [Conics and Perspectivity](chapter-10-conics-and-perspectivity/00-index.ipynb) - Conics through five points, conic cross-ratios, perspective generation, transformations, transfer principles, Pascal and Brianchon.
- Chapter 11: [Calculating with Conics](chapter-11-calculating-with-conics/00-index.ipynb) - Splitting degenerate conics, branching operations, conic-line intersections, conic-conic intersections, complex numbers, and tangent constraints.
- Chapter 12: [Projective d-space](chapter-12-projective-d-space/00-index.ipynb) - Elements at infinity, homogeneous transformations, points and planes in 3-space, lines in 3-space, and universal join/meet systems.
- Chapter 13: [Diagram Techniques](chapter-13-diagram-techniques/00-index.ipynb) - From points, lines, and matrices to tensors, tensor diagrams, transformations, delta tensors, epsilon tensors, and diagrammatic invariants.
- Chapter 14: [Working with Diagrams](chapter-14-working-with-diagrams/00-index.ipynb) - Trace conditions, Pascal's theorem, closed epsilon cycles, conics, RP3 diagrams, rank-4 epsilon-delta rules, and Plucker coordinates.
- Chapter 15: [Configurations, Theorems, and Bracket Expressions](chapter-15-configurations-theorems-and-bracket-expressions/00-index.ipynb) - Desargues theorem, binomial proofs, chains and cycles of cross-ratios, Ceva and Menelaus, and glued configurations.
