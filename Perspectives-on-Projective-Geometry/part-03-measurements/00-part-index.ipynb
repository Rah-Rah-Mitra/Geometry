# Part 3: Measurements

Complex projective line, Euclidean structures, Cayley-Klein geometry, hyperbolic models, and extensions.

## Chapters

- Chapter 16: [Complex Numbers: A Primer](chapter-16-complex-numbers-a-primer/00-index.ipynb) - Historical motivation, the fundamental theorem, geometry of complex numbers, Euler formula, and complex conjugation.
- Chapter 17: [The Complex Projective Line](chapter-17-the-complex-projective-line/00-index.ipynb) - CP1, geometric property tests, projective transformations, inversions, Mobius reflections, Grassmann-Plucker relations, angles, and stereographic projection.
- Chapter 18: [Euclidean Geometry](chapter-18-euclidean-geometry/00-index.ipynb) - Circular points, cocircularity, robust cross-ratios, Euclidean transformations, theorem translation, angles, Laguerre formula, and distances.
- Chapter 19: [Euclidean Structures from a Projective Perspective](chapter-19-euclidean-structures-from-a-projective-perspective/00-index.ipynb) - Mirror images, angle bisectors, circle centers, conic foci, conic construction from foci, triangle theorems, and hybrid thinking.
- Chapter 20: [Cayley-Klein Geometries](chapter-20-cayley-klein-geometries/00-index.ipynb) - I and J revisited, measurements from a conic, nondegenerate and degenerate line measurements, planar Cayley-Klein geometry, and geometry classification.
- Chapter 21: [Measurements and Transformations](chapter-21-measurements-and-transformations/00-index.ipynb) - Measurements versus oriented measurements, transformations, eliminating reference points, comparing measurements, reflections, pole/polar pairs, and rotations.
- Chapter 22: [Cayley-Klein Geometries at Work](chapter-22-cayley-klein-geometries-at-work/00-index.ipynb) - Orthogonality, constructive versus implicit representations, shared and differing properties, midpoints, angle bisectors, and trigonometry.
- Chapter 23: [Circles and Cycles](chapter-23-circles-and-cycles/00-index.ipynb) - Circles via distances, relation to the fundamental conic, centers at infinity, organizing principles, and cycles in Galilean geometry.
- Chapter 24: [Non-Euclidean Geometry: A Historical Interlude](chapter-24-non-euclidean-geometry-a-historical-interlude/00-index.ipynb) - Inner geometry, Euclid's postulates, Gauss, Bolyai, Lobachevsky, Beltrami, Klein, the Beltrami-Klein model, and Poincare.
- Chapter 25: [Hyperbolic Geometry](chapter-25-hyperbolic-geometry/00-index.ipynb) - Hyperbolic staging ground, transformations, angles and boundaries, Poincare disk, CP1 transformations, and distance/angle formulas.
- Chapter 26: [Selected Topics in Hyperbolic Geometry](chapter-26-selected-topics-in-hyperbolic-geometry/00-index.ipynb) - Circles and cycles in the Poincare disk, area and angle defect, Thales and Pythagoras, regular n-gons, and symmetry groups.
- Chapter 27: [What We Did Not Touch](chapter-27-what-we-did-not-touch/00-index.ipynb) - Algebraic projective geometry, discrete mathematics, quantum theory, and dynamic projective geometry.
