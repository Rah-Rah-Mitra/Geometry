# Part 1: Projective Geometry

Projective planes, homogeneous coordinates, cross-ratios, determinants, and bracket algebra.

## Chapters

- Chapter 2: [Projective Planes](chapter-02-projective-planes/00-index.ipynb) - Perspective drawings, incidence axioms, duality, and the smallest finite projective plane.
- Chapter 3: [Homogeneous Coordinates](chapter-03-homogeneous-coordinates/00-index.ipynb) - RP2 from R3, elements at infinity, joins/meets, duality, transformations, and finite projective planes.
- Chapter 4: [Lines and Cross-Ratios](chapter-04-lines-and-cross-ratios/00-index.ipynb) - Coordinates on a projective line, real projective line charts, cross-ratios, and harmonic quadruples.
- Chapter 5: [Calculating with Points on Lines](chapter-05-calculating-with-points-on-lines/00-index.ipynb) - Harmonic points, projective scales, reconstructing real numbers, von Staudt constructions, and Pappos arithmetic.
- Chapter 6: [Determinants](chapter-06-determinants/00-index.ipynb) - Determinantal viewpoint, bracket formulas, Plucker's mu, invariant properties, and Grassmann-Plucker relations.
- Chapter 7: [More on Bracket Algebra](chapter-07-more-on-bracket-algebra/00-index.ipynb) - Moving between points and determinants, a glimpse of invariant theory, invariant functions, and the bracket algebra.
