# Geometric Deep Learning - Standalone Notebook Course

This course is an original executable companion to *Geometric Deep Learning: Grids, Groups, Graphs, Geodesics, and Gauges*. It is designed to stand on its own: each chapter explains the concepts in fresh prose, generates its own visual artifacts, and checks the core invariants numerically.

The PDF is used only as source orientation for structure, concepts, page spans, and notation. The course does not copy textbook prose, exercises, screenshots, page crops, or figures.

## Front Matter Route

- Preface idea: geometric deep learning treats architecture design as the search for invariants and equivariants of structured domains.
- Notation idea: a domain `Omega`, signals `x: Omega -> C`, a symmetry group `G`, a representation `rho(g)`, and a learned map `f` are the recurring objects.
- Body map: Chapter 2 explains why generic high-dimensional learning is hard; Chapters 3-5 build the geometric recipe; Chapters 6-7 show applications and historical roots.
