# Geometry: Euclid and Beyond - Standalone Notebook Course

This course is an original executable notebook edition of Robin Hartshorne's *Geometry: Euclid and Beyond*. It uses the local PDF only for source orientation and replaces passive reading with generated diagrams, plots, model checks, and computational labs.

## Source Notes

- The PDF is image-based; source orientation uses rendered pages rather than text extraction.
- Arabic printed pages map to physical PDF pages by adding 12.
- Back matter is inventoried but is not a canonical teaching notebook.

## Course Units

- [Introduction: Introduction](introduction/00-index.ipynb) - [canonical notebook](introduction/introduction.ipynb); printed pp. 1-6; PDF pp. 13-18; Frames Euclid as the starting point for modern foundations, algebraic models, constructibility, non-Euclidean geometry, and polyhedra.
- [Chapter 1: Euclid's Geometry](chapter-01-euclids-geometry/00-index.ipynb) - [canonical notebook](chapter-01-euclids-geometry/01-euclids-geometry.ipynb); printed pp. 7-64; PDF pp. 19-76; Turns Euclid's constructions and proof style into inspectable construction steps, dependencies, and existence checks.
- [Chapter 2: Hilbert's Axioms](chapter-02-hilberts-axioms/00-index.ipynb) - [canonical notebook](chapter-02-hilberts-axioms/02-hilberts-axioms.ipynb); printed pp. 65-116; PDF pp. 77-128; Shows how incidence, order, congruence, continuity, and line-circle interaction become testable model properties.
- [Chapter 3: Geometry over Fields](chapter-03-geometry-over-fields/00-index.ipynb) - [canonical notebook](chapter-03-geometry-over-fields/03-geometry-over-fields.ipynb); printed pp. 117-164; PDF pp. 129-176; Compares coordinate planes over different algebraic systems and tracks which geometric behaviors require order or completeness.
- [Chapter 4: Segment Arithmetic](chapter-04-segment-arithmetic/00-index.ipynb) - [canonical notebook](chapter-04-segment-arithmetic/04-segment-arithmetic.ipynb); printed pp. 165-194; PDF pp. 177-206; Builds arithmetic operations out of segment constructions and uses similar triangles to introduce coordinates.
- [Chapter 5: Area](chapter-05-area/00-index.ipynb) - [canonical notebook](chapter-05-area/05-area.ipynb); printed pp. 195-240; PDF pp. 207-252; Treats area and volume as invariants tested by dissection, approximation, and obstruction.
- [Chapter 6: Construction Problems and Field Extensions](chapter-06-construction-problems-and-field-extensions/00-index.ipynb) - [canonical notebook](chapter-06-construction-problems-and-field-extensions/06-construction-problems-and-field-extensions.ipynb); printed pp. 241-294; PDF pp. 253-306; Connects constructible geometry to field extensions, polynomial degree, and the boundary between compass-straightedge and marked-ruler power.
- [Chapter 7: Non-Euclidean Geometry](chapter-07-non-euclidean-geometry/00-index.ipynb) - [canonical notebook](chapter-07-non-euclidean-geometry/07-non-euclidean-geometry.ipynb); printed pp. 295-434; PDF pp. 307-446; Uses neutral and hyperbolic models to make the parallel postulate, inversion, Poincare geodesics, ends, and hyperbolic trigonometry inspectable.
- [Chapter 8: Polyhedra](chapter-08-polyhedra/00-index.ipynb) - [canonical notebook](chapter-08-polyhedra/08-polyhedra.ipynb); printed pp. 435-480; PDF pp. 447-492; Studies polyhedra through meshes, Euler characteristic, rigidity intuition, and symmetry diagnostics.
- [Appendix: Brief Euclid](appendix-brief-euclid/00-index.ipynb) - [canonical notebook](appendix-brief-euclid/appendix-brief-euclid.ipynb); printed pp. 481-486; PDF pp. 493-498; Provides a compact generated map of Euclid's book-level structure and proposition themes without reproducing the appendix.
