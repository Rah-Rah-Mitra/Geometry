## Chapter Question

When is a contact manifold the boundary of a symplectic manifold, and what does convexity mean from the symplectic side? Chapter 5 introduces weak and strong fillings, symplectic cobordisms, and the relationship between contact convexity and Levi pseudoconvexity. The chapter is short but structurally important: fillings and caps bridge contact topology with four-dimensional and symplectic arguments.

The notebook uses a collar model. In a strong filling, a Liouville vector field points through the boundary and the contact form is induced by contraction with the symplectic form. In the symplectization picture, `d(e^t alpha)` records this collar explicitly. The visual emphasizes direction: a filling is a boundary with preferred outward symplectic behavior, not merely a manifold that happens to have an edge.


## Translation Guide

A strong filling supplies a symplectic form and a Liouville vector field transverse to the boundary; the induced boundary form is contact. A weak filling relaxes the exact collar requirement but still asks the symplectic form to be positive on the contact planes. A symplectic cobordism has negative and positive boundary components, so it should be read as a directed transition. Levi pseudoconvexity is the complex-geometric cousin of contact-type convexity: both express positivity on complex or contact tangencies. The symplectization coordinate `t` makes scaling visible through `d(e^t alpha)=e^t(dt wedge alpha + d alpha)`.

The checks are symbolic because this chapter is driven by form identities and positivity tests. The visual prevents boundary orientation and Liouville transversality from disappearing into notation.

## Symbolic Check: Symplectization Terms and Levi Positivity

For `alpha=dz+x dy`, the symplectization form has three visible terms: `dt wedge dz`, `x dt wedge dy`, and `dx wedge dy`, all scaled by `e^t`. On a boundary slice, the contact planes inherit the `d alpha` part. For the complex unit ball model, the Levi form of `rho=|z_1|^2+|z_2|^2-1` is positive definite. The JSON check stores these model facts as small algebraic tests.


```python
t,x,y,z=sp.symbols("t x y z")
symplectization_terms={"dt_wedge_dz":str(sp.exp(t)),"dt_wedge_dy":str(x*sp.exp(t)),"dx_wedge_dy":str(sp.exp(t))}
levi_matrix=sp.eye(2)
checks={"symplectization_terms":symplectization_terms,"levi_eigenvalues_unit_ball_model":[int(v) for v in levi_matrix.eigenvals().keys()],"levi_positive_definite":bool(all(v>0 for v in levi_matrix.eigenvals().keys()))}
json_path=save_json(checks,UNIT,"checks","filling-convexity-symbolic-checks.json"); CHECKS.append(json_path); checks
```

## Takeaways

Fillings are contact boundaries plus symplectic interior control. Strong fillings have a Liouville collar that directly induces the contact form; weak fillings ask for positivity on contact planes without the same exact local model. Symplectic cobordisms should be read as directed transitions between contact boundaries. Levi pseudoconvexity gives the complex-geometry version of the same positivity story. These notions become powerful because they allow contact-topological questions to move into symplectic topology and back.
