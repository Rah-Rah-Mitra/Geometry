## Chapter Question

How does a knot remember the contact planes it meets? Chapter 3 introduces Legendrian knots, which are tangent to the contact structure, and transverse knots, which cut through it. It then develops front and Lagrangian projections, approximation theorems, linking numbers, and the classical invariants. This notebook focuses on the front projection because it is the most efficient visual language for Legendrian knots in the standard contact model.

For the form `alpha=dz-y dx`, a Legendrian curve satisfies `dz/dt - y dx/dt = 0`. Away from cusps, the missing coordinate is therefore `y=dz/dx`. A front diagram is not a shadow that forgets the contact geometry; it is a compressed encoding of the Legendrian lift.


## Translation Guide

Legendrian means `alpha(gamma'(t))=0` along the curve. Transverse means `alpha(gamma'(t))` stays nonzero with a chosen sign. A front projection draws `(x,z)` and reconstructs `y` as the slope; cusps occur where that projection has a controlled singularity while the lifted Legendrian curve remains smooth. A Lagrangian projection draws `(x,y)` and makes crossings and area-type data visible. The Thurston-Bennequin invariant can be read from a front as `writhe - half the number of cusps`, and the rotation number is a signed cusp count once orientation conventions are fixed.

The code uses a standard Legendrian unknot front. It is deliberately simple because the goal is to connect diagrammatic rules to the contact equation and invariant formulas.

## Visual 1: A Front and Its Lagrangian Lift

The parametrized front `x=sin(t)`, `z=cos(t)^2` has two cusps. For `alpha=dz-y dx`, the Legendrian lift has `y=dz/dx=-2 sin(t)`, an expression that extends smoothly through the parametrization. The left panel is the front. The right panel is the associated `(x,y)` projection.

The inspection target is the way slope becomes a coordinate. Steep parts of the front correspond to larger absolute `y`; the cusp is not a failure of the Legendrian curve, but a feature of this projection. This is why front diagrams support so much low-dimensional contact topology.


```python
t=np.linspace(0,2*np.pi,900,endpoint=True)
x_front=np.sin(t); z_front=np.cos(t)**2; y_lift=-2*np.sin(t)
alpha_residual=np.gradient(z_front,t)-y_lift*np.gradient(x_front,t)
fig,axes=plt.subplots(1,2,figsize=(10,4.2))
axes[0].plot(x_front,z_front,color="#2f6f9f",linewidth=2.2)
axes[0].scatter([1,-1],[0,0],color="#c84630",zorder=3,label="cusps")
axes[0].set_title("Front projection (x,z)"); axes[0].set_xlabel("x"); axes[0].set_ylabel("z"); axes[0].set_aspect("equal"); axes[0].legend(loc="upper right")
axes[1].plot(x_front,y_lift,color="#6b5b95",linewidth=2.2)
axes[1].set_title("Lagrangian projection (x,y)"); axes[1].set_xlabel("x"); axes[1].set_ylabel("y = dz/dx"); axes[1].set_aspect("equal")
front_path=save_matplotlib(fig,UNIT,"figures","legendrian-front-and-lagrangian-projection.png")
plt.close(fig); ARTIFACTS.append(front_path); display_artifact(front_path,width=820)
```

## Classical Invariant Check

For this front, the crossing writhe is zero and there are two cusps, one of each orientation in the convention used here. The front formulas give `tb=-1` and rotation number `0`. These numbers are not decorative labels: they are Legendrian isotopy invariants and later become framing input for contact surgery. The JSON check also records the maximum residual in the contact equation along the sampled lift. That residual is the bridge from the picture to the actual tangency condition.


```python
invariants=legendrian_front_invariants(writhe=0,cusps_up=1,cusps_down=1)
checks={**invariants,"max_abs_alpha_on_velocity":float(np.max(np.abs(alpha_residual))),"model":"x=sin(t), z=cos(t)^2, y=-2 sin(t)"}
json_path=save_json(checks,UNIT,"checks","legendrian-unknot-front-invariants.json")
CHECKS.append(json_path); checks
```

## Takeaways

Legendrian and transverse knots are ordinary knots plus contact-plane incidence data. Projections make that extra data computable. A front diagram stores the missing coordinate as slope, and the simplest classical invariants can be read from crossings and cusps. Linking enters because push-offs compare the knot with a nearby copy determined by a framing. These ideas become essential in the surgery chapter, where a topological attachment coefficient must be measured relative to contact framing rather than an arbitrary diagrammatic convention.
