## Chapter Question

What does it mean for a hyperplane field to be as far from integrable as possible, and why does that condition appear in mechanics, first-order differential equations, wavefronts, and topology? This notebook treats the opening chapter as a laboratory for the objects that can be inspected directly: contact forms, contact planes, Reeb directions, and contact elements.

The working model is the standard three-dimensional form `alpha = dz + x dy`. In an integrable plane field, nearby tangent planes assemble into surfaces. In this model the planes twist just enough that no surface can have all its tangent planes equal to the field. That twist is not a slogan; the coefficient of `alpha wedge d alpha` is a computable obstruction to Frobenius integrability.


## Translation Guide

The source chapter moves among differential forms, projectivized cotangent bundles, mechanics, geodesic flow, and applications to topology. Computationally, each language becomes a small check. A hyperplane field is the kernel of a one-form. The contact condition is the nonzero top-degree coefficient of `alpha wedge (d alpha)^n`; in dimension three that is the coefficient of `dx wedge dy wedge dz`. The Reeb vector field is the unique vector satisfying `alpha(R)=1` and `d alpha(R,-)=0`. A contact element is first-order tangent-hyperplane data, so the space of contact elements is a natural home for slopes, first-order differential equations, and contact transformations.

The visuals stay local because that is where the first habit forms: look at the moving planes, compute the volume coefficient, and name the complementary Reeb direction.

## Visual 1: Standard Contact Planes

The figure samples two spanning directions of `ker(dz + x dy)`: the horizontal vector `partial_x` and the tilted vector `partial_y - x partial_z`. At `x=0` the second vector is horizontal; for positive and negative `x` it tilts in opposite vertical directions. The inspection target is the shear across the `x` direction. This is the visible trace of non-integrability: the planes do not stack into leaves.

A quiver plot is better than a fake surface here, because the theorem says the integral surface should not exist. We inspect the distribution itself.


```python
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401
xs = np.linspace(-1.2, 1.2, 5)
ys = np.linspace(-1.0, 1.0, 5)
X, Y = np.meshgrid(xs, ys)
Z = np.zeros_like(X)
fig = plt.figure(figsize=(8.5, 6.5))
ax = fig.add_subplot(111, projection="3d")
ax.quiver(X, Y, Z, 0.42, 0, 0, color="#2f6f9f", length=0.45, normalize=False, label="partial_x")
ax.quiver(X, Y, Z, 0, 0.42, -0.42 * X, color="#c84630", length=0.55, normalize=False, label="partial_y - x partial_z")
ax.set_title("Sampled planes in ker(dz + x dy)")
ax.set_xlabel("x"); ax.set_ylabel("y"); ax.set_zlabel("z")
ax.set_zlim(-1.0, 1.0)
ax.view_init(elev=23, azim=-54)
plane_path = save_matplotlib(fig, UNIT, "figures", "standard-contact-plane-field.png")
plt.close(fig)
ARTIFACTS.append(plane_path)
display_artifact(plane_path, width=760)
```

## Contact Volume and Reeb Direction

For `alpha = dz + lambda x dy`, the contact-volume coefficient is `lambda`. The case `lambda=0` collapses to `dz`, whose kernel integrates to horizontal planes. Every nonzero value restores maximal non-integrability, with the sign recording orientation. For the standard form, `d alpha = dx wedge dy`; the vector `partial_z` is killed by `d alpha` and evaluates to one under `alpha`, so it is the Reeb direction.

The chapter later warns that a Reeb field belongs to a chosen contact form, not only to the contact structure. Scaling the form may preserve the kernel while changing the Reeb dynamics. The small parameter experiment below is the first place where that distinction is visible.


```python
x, y, z, lam = sp.symbols("x y z lambda")
volume = contact_volume_coefficient_3d(0, lam * x, 1, (x, y, z))
checks = {
    "alpha": "dz + lambda*x*dy",
    "contact_volume_coefficient": str(volume),
    "standard_lambda_1": str(volume.subs(lam, 1)),
    "integrable_lambda_0": str(volume.subs(lam, 0)),
    "reeb_for_lambda_nonzero": "partial_z",
}
json_path = save_json(checks, UNIT, "checks", "contact-volume-and-reeb.json")
CHECKS.append(json_path)
checks
```

## Applied Lab

Vary `lambda` in `dz + lambda x dy`. The bar chart records the signed coefficient of `alpha wedge d alpha`. The zero bar is not a small contact structure; it is failure of the contact condition. Negative values are still contact forms, but they reverse the induced orientation relative to positive values. This is a compact way to see why contactness is an open nonvanishing condition while the exact coefficient depends on coordinates and on the chosen defining form.

## Takeaways

A contact structure is a hyperplane field with a built-in failure of Frobenius integrability. In dimension three, the failure is exactly the nonzero coefficient of `alpha wedge d alpha`. The Reeb vector field is the complementary direction selected by a particular contact form. Spaces of contact elements explain the name: a point plus a tangent hyperplane is first-order contact data, and contact transformations preserve the geometry of that data. The rest of the course repeatedly returns to this local model while adding global topology, knots, surfaces, fillings, and surgery.
