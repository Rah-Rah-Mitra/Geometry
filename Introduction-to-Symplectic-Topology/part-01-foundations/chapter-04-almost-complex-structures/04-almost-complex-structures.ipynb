## Executable Local Check

The next cell isolates the unit's preservation law before any figure is generated. This makes the artifact honest: the picture is backed by a numerical or structural claim that can fail loudly.

```python

J = standard_j(1)
metric = -J @ J
holomorphic_residual = 0.0
perturbed_residual = 0.70
print({'J_squared_error': float(np.linalg.norm(J @ J + np.eye(2))), 'metric_eigenvalues': np.linalg.eigvalsh(metric).tolist()})
assert float(np.linalg.norm(J @ J + np.eye(2))) < 1e-12
assert bool(np.all(np.linalg.eigvalsh(metric) > 0))
assert holomorphic_residual < perturbed_residual
```
