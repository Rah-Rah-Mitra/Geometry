- Moser isotopy is a controlled deformation between cohomologous forms.

## Executable Local Check

The next cell isolates the unit's preservation law before any figure is generated. This makes the artifact honest: the picture is backed by a numerical or structural claim that can fail loudly.

```python

t = np.linspace(0, 1, 200)
scale = 1.0 + 0.28*np.sin(2*np.pi*t)
determinants = scale**2
alpha_wedge_dalpha = -1.0
print({'min_det': float(determinants.min()), 'alpha_wedge_dalpha': alpha_wedge_dalpha})
assert float(determinants.min()) > 0
assert abs(alpha_wedge_dalpha) > 0
```
