## Executable Local Check

The next cell isolates the unit's preservation law before any figure is generated. This makes the artifact honest: the picture is backed by a numerical or structural claim that can fail loudly.

```python

R = 1.2
r = 1.0
lam = 1.7
ball_capacity = math.pi * R**2
cylinder_capacity = math.pi * r**2
print({'ball_capacity': ball_capacity, 'cylinder_capacity': cylinder_capacity, 'scaled_capacity': math.pi*(lam*R)**2})
assert ball_capacity > cylinder_capacity
assert abs(math.pi*(lam*R)**2 - lam**2 * ball_capacity) < 1e-12
```
