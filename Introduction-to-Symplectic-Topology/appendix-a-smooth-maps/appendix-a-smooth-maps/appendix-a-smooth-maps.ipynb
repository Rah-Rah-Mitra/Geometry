## Executable Local Check

The next cell isolates the unit's preservation law before any figure is generated. This makes the artifact honest: the picture is backed by a numerical or structural claim that can fail loudly.

```python

xs = np.linspace(-0.25, 1.25, 500)
step = smooth_step(xs)
bump = smooth_step(xs) * smooth_step(1-xs)
derivative = finite_difference(step, xs)
left = float(abs(derivative[np.argmin(abs(xs-0.0))]))
right = float(abs(derivative[np.argmin(abs(xs-1.0))]))
print({'left_derivative': left, 'right_derivative': right, 'support_outside': float(np.max(np.abs(bump[(xs <= 0) | (xs >= 1)])))})
assert left < 5e-3 and right < 5e-3
assert float(np.max(np.abs(bump[(xs <= 0) | (xs >= 1)]))) < 1e-12
```
