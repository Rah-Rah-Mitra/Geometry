## Executable Local Check

The next cell isolates the unit's preservation law before any figure is generated. This makes the artifact honest: the picture is backed by a numerical or structural claim that can fail loudly.

```python

eps = 0.34
original = np.array([[0,0], [1,0], [1,1], [0,1]])
blown = np.array([[eps,0], [1,0], [1,1], [0,1], [0,eps]])
loss = abs(polygon_signed_area(original)) - abs(polygon_signed_area(blown))
print({'area_loss': loss, 'expected': 0.5*eps**2})
assert abs(loss - 0.5*eps**2) < 1e-12
```
