## Executable Local Check

The next cell isolates the unit's preservation law before any figure is generated. This makes the artifact honest: the picture is backed by a numerical or structural claim that can fail loudly.

```python
coupling = 0.35
critical = np.array([[-np.pi, -np.pi], [0.0, 0.0], [np.pi, np.pi]])
def grad(point):
    a, b = point
    return np.array([-(b-a) - coupling*np.sin(a), (b-a) - coupling*np.sin(b)])
residuals = [float(np.linalg.norm(grad(point))) for point in critical]
print({'critical_points': critical.tolist(), 'gradient_residuals': residuals})
assert max(residuals) < 1e-10
assert len(critical) >= 3
```
