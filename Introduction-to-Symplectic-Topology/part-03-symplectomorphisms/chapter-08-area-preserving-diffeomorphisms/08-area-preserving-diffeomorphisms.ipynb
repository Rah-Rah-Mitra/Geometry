## Executable Local Check

The next cell isolates the unit's preservation law before any figure is generated. This makes the artifact honest: the picture is backed by a numerical or structural claim that can fail loudly.

```python

r = np.linspace(1, 2, 50)
twist = 1.6 * (r - 1.5)
jacobian = np.array([[1.0, 0.0], [1.6, 1.0]])
print({'inner_twist': float(twist[0]), 'outer_twist': float(twist[-1]), 'det': float(np.linalg.det(jacobian))})
assert float(twist[0] * twist[-1]) < 0
assert abs(float(np.linalg.det(jacobian)) - 1.0) < 1e-12
```
