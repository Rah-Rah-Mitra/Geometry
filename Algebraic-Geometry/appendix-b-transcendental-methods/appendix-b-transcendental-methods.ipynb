## Appendix Goal

This appendix explains how algebraic varieties over the complex numbers talk to complex analytic spaces and compact complex manifolds. ## Translation Guide

A scheme of finite type over the complex numbers is translated into an analytic space by reading polynomial equations as holomorphic equations and gluing the resulting analytic pieces. Coherent algebraic sheaves are translated into coherent analytic sheaves. In the projective case, GAGA says this translation preserves enough information to recover sheaves, cohomology, and morphisms. Compact complex manifolds are then tested for algebraicity by looking at meromorphic functions, projectivity, Kahler forms, Hodge classes, and Moishezon behavior. The exponential sequence translates additive holomorphic data into multiplicative line-bundle data and sends Picard information toward integral cohomology.

The exponential sequence is the appendix's most concrete transcendental tool. Additive holomorphic functions exponentiate to nowhere-vanishing multiplicative functions, and the kernel is integral. Passing to cohomology places the Picard group between analytic additive data and integral second cohomology. That is why the first Chern class appears here, and why the Neron-Severi and Picard varieties can be studied using analytic topology. The winding lab shows the basic exactness: integer shifts in the additive coordinate return to the same multiplicative value.


```python
a, b = 0.23 + 0.14j, -0.41 + 0.07j
exp_residual = abs(cmath.exp(2j*math.pi*(a+b)) - cmath.exp(2j*math.pi*a) * cmath.exp(2j*math.pi*b))
integer_shift_residual = max(abs(cmath.exp(2j*math.pi*(0.37+n)) - cmath.exp(2j*math.pi*0.37)) for n in range(-3, 4))
comparison_rows = [
    {"comparison": "coherent sheaves", "projective_case": "equivalent"},
    {"comparison": "cohomology dimensions", "projective_case": "matched"},
    {"comparison": "morphisms to projective space", "projective_case": "algebraizable"},
    {"comparison": "compact analytic subspace of projective space", "projective_case": "algebraic by Chow"},
]
source_table_path = save_table(source_sections, ARTIFACT_ROOT, "tables", "appendix-b-source-coverage.csv")
comparison_table_path = save_table(comparison_rows, ARTIFACT_ROOT, "tables", "gaga-comparison-ledger.csv")
checks_path = save_json({
    "hermitian_eigenvalues": [float(v) for v in eigenvalues],
    "hermitian_positive_definite": bool(np.all(eigenvalues > 0)),
    "exponential_additive_multiplicative_residual": exp_residual,
    "integer_shift_residual": integer_shift_residual,
    "comparison_rows": comparison_rows,
}, ARTIFACT_ROOT, "checks", "transcendental-methods-checks.json")
generated_artifacts.extend([source_table_path, comparison_table_path, checks_path])
assert np.all(eigenvalues > 0)
assert exp_residual < 1e-12
assert integer_shift_residual < 1e-12
assert_artifacts(generated_artifacts)
records = []
for artifact in generated_artifacts:
    record = {"path": relpath(artifact), "exists": artifact.exists(), "bytes": artifact.stat().st_size}
    if artifact.suffix.lower() == ".png": record.update(image_stats(artifact))
    records.append(record)
final_sanity = {
    "appendix": "Appendix B: Transcendental Methods",
    "source_span": {"printed": "438-448", "pdf": "453-463"},
    "artifacts": records,
    "checks": {
        "hermitian_positive_definite": True,
        "exponential_sequence_residuals_zero": True,
        "gaga_comparison_ledger_present": True,
        "source_sections_covered": len(source_sections),
    },
    "standalone_contract": "original prose, generated visuals, and local checks; no copied source text or figures",
}
final_path = save_json(final_sanity, ARTIFACT_ROOT, "checks", "final-sanity.json")
final_sanity["artifacts"].append({"path": relpath(final_path), "exists": final_path.exists(), "bytes": final_path.stat().st_size})
save_json(final_sanity, ARTIFACT_ROOT, "checks", "final-sanity.json")
final_sanity
```

## Standalone Study Notes

Appendix B is best read as a dictionary with warnings attached. The associated analytic space construction is straightforward locally: polynomial equations are holomorphic equations. The subtle point is not making an analytic object, but knowing when the analytic object remembers the algebraic one. GAGA gives a strong yes in the projective setting: coherent sheaves, cohomology, and morphisms can be compared. Outside that setting, compact complex manifolds and analytic spaces can behave in ways that are not captured by schemes alone.

The algebraicity criteria section organizes several different tests. A compact Riemann surface is algebraic, but higher-dimensional compact complex manifolds need more structure. Moishezon means there are enough meromorphic functions, Kahler means there is a closed positive `(1,1)` form, and Hodge conditions connect the form to integral cohomology. The implication chart should be read as a map of hypotheses, not as a complete proof. The exponential sequence then gives a concrete analytic tool: additive holomorphic functions, multiplicative invertible functions, Picard classes, and integral cohomology sit in one long exact sequence. That sequence is the computational doorway to Picard varieties, Neron-Severi groups, and Chern classes over the complex numbers.
