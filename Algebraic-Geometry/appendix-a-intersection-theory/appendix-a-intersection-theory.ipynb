## Appendix Goal

This appendix generalizes the surface intersection pairing into a higher-dimensional bookkeeping system. The main objects are cycles, rational equivalence, Chow rings, Chern classes, and Riemann-Roch formulas. The source is brief, but it carries a lot of information: it explains why divisors alone are not enough in higher dimension, why intersections of cycles require moving and local multiplicities, why Chern classes encode vector bundles, and why Hirzebruch-Riemann-Roch packages Euler characteristics as top-degree intersections.

## Translation Guide

A cycle is represented as a formal integer combination of subvarieties of fixed codimension. Rational equivalence is represented as the relation that lets cycles move through a one-parameter family without changing their Chow class. The Chow ring is represented as a graded multiplication table. Chern classes are represented by formal roots and total Chern polynomials. Riemann-Roch is represented as a top-degree extraction from a product of a Chern character and a Todd class.

The pipeline diagram shows why the appendix begins with cycles rather than only divisors. In dimension greater than two, intersecting divisors may produce codimension-two cycles, and the product must live in a graded group. Rational equivalence is the relation that makes movement legitimate; the diagonal converts products into pullback and pushforward language. The diagram is a proof scaffold: it tells the learner which operations must be compatible for a Chow ring to exist.


```python
n = 3
powers = list(range(n+1))
product = np.full((n+1, n+1), -1, dtype=int)
for i in powers:
    for j in powers:
        if i + j <= n:
            product[i, j] = i + j
fig, ax = plt.subplots(figsize=(6.8, 5.8))
im = ax.imshow(product, cmap="YlGnBu", vmin=-1, vmax=n)
ax.set_xticks(powers, [f"h^{j}" if j else "1" for j in powers])
ax.set_yticks(powers, [f"h^{i}" if i else "1" for i in powers])
for i in powers:
    for j in powers:
        label = "0" if product[i, j] < 0 else ("1" if product[i,j] == 0 else f"h^{product[i,j]}")
        ax.text(j, i, label, ha="center", va="center", color="black")
ax.set_title("Chow ring A(P3)=Z[h]/(h^4)")
chow_path = save_matplotlib(fig, ARTIFACT_ROOT, "figures", "projective-space-chow-multiplication-table.png")
plt.close(fig)
generated_artifacts.append(chow_path)
display_artifact(chow_path, width=620)
```

The projective-space multiplication table is the smallest useful Chow-ring laboratory. The hyperplane class `h` generates the ring, multiplication adds codimension, and powers beyond the dimension vanish. The top class `h^3` has degree one on `P^3`; this is the model behind top-degree extraction in Riemann-Roch. The table also makes clear why an intersection product in higher dimension is not automatically a number: only the top codimension can be pushed to an integer by the degree map.


```python
t, a, b, c = sp.symbols("t a b c")
cE = sp.expand((1 + a*t) * (1 + b*t))
cF = 1 + c*t
cTotal = sp.expand(cE * cF)
chern_coeffs = [sp.expand(cTotal).coeff(t, k) for k in range(4)]
fig, axes = plt.subplots(1, 2, figsize=(11, 4.6))
coeff_labels = ["c0", "c1", "c2", "c3"]
axes[0].bar(coeff_labels, [1, 3, 3, 1], color=["#1971c2", "#2b8a3e", "#f08c00", "#9d0208"])
axes[0].set_title("Splitting-principle coefficient slots")
axes[0].set_ylabel("rank-3 binomial model")
axes[1].axis("off")
text = "Exact sequence check:\n"
text += "c(E) = (1+a t)(1+b t)\n"
text += "c(F) = (1+c t)\n"
text += "c(E) c(F) = " + str(cTotal) + "\n\n"
text += "coefficients:\n" + "\n".join(f"{label}: {expr}" for label, expr in zip(coeff_labels, chern_coeffs))
axes[1].text(0, .95, text, va="top", family="monospace", fontsize=11)
axes[1].set_title("Total Chern classes multiply")
chern_path = save_matplotlib(fig, ARTIFACT_ROOT, "figures", "chern-class-splitting-dashboard.png")
plt.close(fig)
generated_artifacts.append(chern_path)
display_artifact(chern_path, width=780)
```

Chern classes translate locally free sheaves into Chow-ring classes. The splitting principle lets a bundle be checked after pulling back to a space where it has formal line-bundle roots. The exact-sequence rule says total Chern classes multiply. This is why Chern classes are the correct replacement for the divisor associated to a line bundle: they encode how vector-bundle data contributes to intersection formulas in every codimension.


```python
degrees = np.arange(-4, 9)
chi_p2 = ((degrees + 2) * (degrees + 1) // 2).astype(int)
fig, ax = plt.subplots(figsize=(9, 4.8))
colors = ["#d9480f" if value < 0 else "#1971c2" for value in chi_p2]
ax.bar(degrees, chi_p2, color=colors)
ax.axhline(0, color="#333", lw=.8)
for d, value in zip(degrees, chi_p2):
    ax.text(d, value + (0.4 if value >= 0 else -0.8), str(int(value)), ha="center", va="center", fontsize=8)
ax.set_title("Hirzebruch-Riemann-Roch ledger for O(d) on P2")
ax.set_xlabel("degree d"); ax.set_ylabel("chi(O(d))")
ax.grid(axis="y", alpha=0.2)
hrr_path = save_matplotlib(fig, ARTIFACT_ROOT, "figures", "hirzebruch-riemann-roch-ledger.png")
plt.close(fig)
generated_artifacts.append(hrr_path)
display_artifact(hrr_path, width=760)

fig_html = go.Figure()
fig_html.add_trace(go.Scatter(x=degrees, y=chi_p2, mode="lines+markers", name="chi(O_P2(d))"))
fig_html.add_trace(go.Scatter(x=degrees, y=(degrees**2)/2, mode="lines", name="leading top-degree term", line={"dash":"dash"}))
fig_html.update_layout(title="Top-degree extraction lab for Riemann-Roch on P2", xaxis_title="d", yaxis_title="Euler characteristic", template="plotly_white", height=500)
lab_path = save_plotly_html(fig_html, ARTIFACT_ROOT, "html", "top-degree-riemann-roch-lab.html")
generated_artifacts.append(lab_path)
display_artifact(lab_path, width="100%", height=410)
```

The Riemann-Roch ledger shows how the appendix's formalism returns to a familiar numerical task. For line bundles on the projective plane, the Euler characteristic is a quadratic polynomial in the degree. The leading term comes from top-degree intersection, and the lower terms come from the Todd correction. In the general theorem, this same pattern is expressed as the degree of the top-dimensional part of `ch(E) td(T_X)`. The interactive lab makes the polynomial behavior visible and separates the leading intersection term from the correction terms.


```python
chow_truncation_ok = bool(np.all(product[np.add.outer(powers, powers) > n] == -1))
chern_exact_residual = sp.simplify(cTotal - cE*cF)
hrr_rows = [{"degree": int(d), "chi": int(v), "binomial_formula": int((d+2)*(d+1)//2)} for d, v in zip(degrees, chi_p2)]
source_table_path = save_table(source_sections, ARTIFACT_ROOT, "tables", "appendix-a-source-coverage.csv")
hrr_table_path = save_table(hrr_rows, ARTIFACT_ROOT, "tables", "p2-riemann-roch-ledger.csv")
checks_path = save_json({
    "chow_ring_h4_zero": chow_truncation_ok,
    "chern_exact_sequence_residual": str(chern_exact_residual),
    "chern_total_coefficients": [str(item) for item in chern_coeffs],
    "p2_riemann_roch_rows": hrr_rows,
}, ARTIFACT_ROOT, "checks", "intersection-theory-checks.json")
generated_artifacts.extend([source_table_path, hrr_table_path, checks_path])
assert chow_truncation_ok
assert chern_exact_residual == 0
assert all(row["chi"] == row["binomial_formula"] for row in hrr_rows)
assert_artifacts(generated_artifacts)
records = []
for artifact in generated_artifacts:
    record = {"path": relpath(artifact), "exists": artifact.exists(), "bytes": artifact.stat().st_size}
    if artifact.suffix.lower() == ".png": record.update(image_stats(artifact))
    records.append(record)
final_sanity = {
    "appendix": "Appendix A: Intersection Theory",
    "source_span": {"printed": "424-437", "pdf": "439-452"},
    "artifacts": records,
    "checks": {
        "chow_ring_truncation_checked": True,
        "chern_exact_sequence_product_checked": True,
        "riemann_roch_p2_formula_checked": True,
        "source_sections_covered": len(source_sections),
    },
    "standalone_contract": "original prose, generated visuals, and local checks; no copied source text or figures",
}
final_path = save_json(final_sanity, ARTIFACT_ROOT, "checks", "final-sanity.json")
final_sanity["artifacts"].append({"path": relpath(final_path), "exists": final_path.exists(), "bytes": final_path.stat().st_size})
save_json(final_sanity, ARTIFACT_ROOT, "checks", "final-sanity.json")
final_sanity
```

## Standalone Study Notes

Appendix A compresses a large theory, so the notebook separates the roles of the main objects. Cycles are the raw formal sums. Rational equivalence is the movement rule that lets one replace a cycle by a better-positioned representative. Chow groups are the resulting graded containers. The Chow ring is the extra structure that turns proper intersections into products. Chern classes are the way vector bundles enter this ring, and the splitting principle is the practical reason formal roots are useful. Riemann-Roch then says that an Euler characteristic can be computed by multiplying characteristic classes and taking the top-dimensional part.

The projective-space example is intentionally simple because it exposes the grading. In `A(P^3)`, multiplying by the hyperplane class increases codimension, and `h^4` vanishes because there is no codimension-four class on a threefold. The Chern dashboard shows why total Chern classes are multiplicative in exact sequences. The Riemann-Roch ledger for `O(d)` on `P^2` gives a familiar polynomial output, so the abstract expression `ch(E) td(T_X)` can be read as an executable recipe rather than a slogan. The complements section points beyond the notebook: positivity, standard conjectures, Grothendieck-Riemann-Roch, and singular variants all reuse the same bookkeeping architecture.
