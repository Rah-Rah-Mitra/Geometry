# Basic Topology - Standalone Notebook Course

This course is an original executable notebook edition of M. A. Armstrong's *Basic Topology*. It uses the local PDFs only for source orientation and replaces passive reading with generated diagrams, plots, algebra checks, and computational labs.

## Source Notes

- The main PDF is text-extractable; printed page 1 starts at physical PDF page 12.
- The solutions PDF is a private cross-check for exercise intent only and is not copied.
- Back matter is inventoried but is not a canonical teaching notebook.

## Course Units

- [Chapter 1: Introduction](chapter-01-introduction/00-index.ipynb) - [canonical notebook](chapter-01-introduction/01-introduction.ipynb); printed pp. 1-26; PDF pp. 12-37; Turns the opening motivation into a laboratory for Euler characteristic, surface examples, and the first invariant-thinking habits.
- [Chapter 2: Continuity](chapter-02-continuity/00-index.ipynb) - [canonical notebook](chapter-02-continuity/02-continuity.ipynb); printed pp. 27-42; PDF pp. 38-53; Recasts continuity as preimage control and uses approximating pictures to separate intuition from theorem statements.
- [Chapter 3: Compactness and Connectedness](chapter-03-compactness-and-connectedness/00-index.ipynb) - [canonical notebook](chapter-03-compactness-and-connectedness/03-compactness-and-connectedness.ipynb); printed pp. 43-64; PDF pp. 54-75; Builds finite-subcover and no-separation instincts with covers, products, and examples where connectedness and paths differ.
- [Chapter 4: Identification Spaces](chapter-04-identification-spaces/00-index.ipynb) - [canonical notebook](chapter-04-identification-spaces/04-identification-spaces.ipynb); printed pp. 65-86; PDF pp. 76-97; Makes quotient constructions visible by drawing gluing instructions, orbit representatives, and the maps that forget labels.
- [Chapter 5: The Fundamental Group](chapter-05-the-fundamental-group/00-index.ipynb) - [canonical notebook](chapter-05-the-fundamental-group/05-the-fundamental-group.ipynb); printed pp. 87-118; PDF pp. 98-128; Treats loops as movable data and turns group operations, winding, and deformation retraction into inspectable computations.
- [Chapter 6: Triangulations](chapter-06-triangulations/00-index.ipynb) - [canonical notebook](chapter-06-triangulations/06-triangulations.ipynb); printed pp. 119-148; PDF pp. 129-158; Uses simple complexes as a computational language for spaces, maps, subdivisions, and edge-loop presentations.
- [Chapter 7: Surfaces](chapter-07-surfaces/00-index.ipynb) - [canonical notebook](chapter-07-surfaces/07-surfaces.ipynb); printed pp. 149-172; PDF pp. 159-181; Builds the classification theorem as a hands-on system of symbols, cuts, handles, crosscaps, and Euler-characteristic checks.
- [Chapter 8: Simplicial Homology](chapter-08-simplicial-homology/00-index.ipynb) - [canonical notebook](chapter-08-simplicial-homology/08-simplicial-homology.ipynb); printed pp. 173-194; PDF pp. 182-202; Converts cycles, boundaries, and holes into chain groups, boundary matrices, ranks, and visible representatives.
- [Chapter 9: Degree and Lefschetz Number](chapter-09-degree-and-lefschetz-number/00-index.ipynb) - [canonical notebook](chapter-09-degree-and-lefschetz-number/09-degree-and-lefschetz-number.ipynb); printed pp. 195-212; PDF pp. 203-220; Connects global map invariants to fixed points, antipodal symmetry, Euler-Poincare counts, and dimension tests.
- [Chapter 10: Knots and Covering Spaces](chapter-10-knots-and-covering-spaces/00-index.ipynb) - [canonical notebook](chapter-10-knots-and-covering-spaces/10-knots-and-covering-spaces.ipynb); printed pp. 213-240; PDF pp. 221-247; Treats knots as embedded circles whose complements, coverings, and algebraic invariants can be computed from diagrams.
- [Appendix: Generators and Relations](appendix-generators-and-relations/00-index.ipynb) - [canonical notebook](appendix-generators-and-relations/appendix-generators-and-relations.ipynb); printed pp. 241-243; PDF pp. 248-250; Supplies the algebraic grammar used by edge groups, knot groups, and presentation matrices.
