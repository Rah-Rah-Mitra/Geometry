# Fundamentals of Computer Graphics - Visualization-First Notebook Course

This is a standalone notebook course built from the chapter structure of *Fundamentals of Computer Graphics*, Fifth Edition. The notebooks use original prose, generated diagrams, executable checks, and book-local artifacts.

## How To Use This Course

Start with any chapter. Each folder contains a local `00-index.ipynb` and one canonical teaching notebook. Generated artifacts live under `artifacts/chapter-XX/`.
