```python
concepts = ["Lattices", "Minkowski's", "Theorem", "Minkowski", "General", "Number", "Theoretic"]
G = nx.DiGraph()
G.add_node(UNIT_TITLE, role="source unit")
for idx, concept in enumerate(concepts):
    G.add_node(concept, role="concept")
    G.add_edge(UNIT_TITLE, concept)
    if idx > 0:
        G.add_edge(concepts[idx - 1], concept)
G.add_node("Invariant check", role="validation")
G.add_node("Learner lab", role="experiment")
G.add_edge(concepts[-1], "Invariant check")
G.add_edge(concepts[max(0, len(concepts)//2 - 1)], "Learner lab")
G.add_edge("Learner lab", "Invariant check")

pos = nx.spring_layout(G, seed=len(UNIT_KEY), k=0.9)
fig, ax = plt.subplots(figsize=(10, 6))
colors = [{"source unit": "#355c7d", "concept": "#6c9a8b", "validation": "#c06c84", "experiment": "#f8b195"}.get(G.nodes[n].get("role"), "#dddddd") for n in G.nodes]
nx.draw_networkx_edges(G, pos, ax=ax, arrows=True, arrowstyle="-|>", width=1.4, edge_color="#555555")
nx.draw_networkx_nodes(G, pos, ax=ax, node_color=colors, node_size=1500, linewidths=1, edgecolors="#222222")
nx.draw_networkx_labels(G, pos, ax=ax, font_size=8)
ax.set_title(f"Dependency map: {UNIT_TITLE}")
ax.axis("off")
concept_map_path = save_matplotlib(fig, FIGURES / "lattices-and-minkowski-s-theorem-dependency-map.png")
plt.close(fig)
concept_summary_path = save_json({"unit": UNIT_TITLE, "concepts": concepts, "edges": list(G.edges())}, CHECKS / "lattices-and-minkowski-s-theorem-dependency-map.json")
display_artifact(concept_map_path, width=820)
```

The dependency map is a proof-reading surface: follow arrows from the source-unit question toward the invariant check. A useful inspection habit is to ask which node would fail first if a hypothesis from the chapter were removed. For `Lattices and Minkowski's Theorem`, the answer should involve one of the concrete terms in the graph rather than a vague appeal to general theory.


```python
from scipy.spatial import ConvexHull, Delaunay
rng = np.random.default_rng(seed=len(UNIT_KEY) * 17)
angles = np.linspace(0, 2*np.pi, 9, endpoint=False)
radii = 1.0 + 0.18*np.sin(np.arange(9) + len(UNIT_TITLE))
pts = np.column_stack([radii*np.cos(angles), radii*np.sin(angles)]) + 0.05 * rng.normal(size=(9, 2))
hull = ConvexHull(pts)
tri = Delaunay(pts)
orient = float(np.linalg.det(np.vstack([pts[1]-pts[0], pts[2]-pts[0]])))
fig, ax = plt.subplots(figsize=(7, 7))
ax.triplot(pts[:,0], pts[:,1], tri.simplices, color="#9aa7b1", lw=1)
loop = np.r_[hull.vertices, hull.vertices[0]]
ax.plot(pts[loop,0], pts[loop,1], color="#c8553d", lw=2.5, label="convex hull")
ax.scatter(pts[:,0], pts[:,1], s=55, color="#2a9d8f", zorder=3)
for i, (px, py) in enumerate(pts):
    ax.text(px+0.025, py+0.025, str(i), fontsize=8)
ax.set_aspect("equal")
ax.set_title("Lattices and Minkowski's Theorem: finite point configuration")
ax.legend()
model_path = save_matplotlib(fig, FIGURES / "lattices-and-minkowski-s-theorem-finite-configuration.png")
plt.close(fig)
model_check_path = save_json({"unit": UNIT_TITLE, "focus": "Minkowski theorem, general lattices, and number-theoretic applications.", "point_count": int(len(pts)), "hull_vertex_count": int(len(hull.vertices)), "simplex_count": int(len(tri.simplices)), "orientation_sample": orient, "nondegenerate_orientation": abs(orient) > 1e-6}, CHECKS / "lattices-and-minkowski-s-theorem-finite-configuration.json")
display_artifact(model_path, width=720)
```

```python
x, y, t = sp.symbols("x y t", real=True)
matrix = sp.Matrix([[1, x, x**2], [1, y, y**2], [1, t, t**2]])
vandermonde = sp.factor(matrix.det())
residual_at_repeat = sp.simplify(vandermonde.subs(t, x))
numeric_vector = np.array([len(c) for c in concepts], dtype=float)
normalized = numeric_vector / numeric_vector.sum()
entropy = float(-(normalized * np.log(normalized)).sum())
invariant_report = {
    "unit": UNIT_TITLE,
    "formulas": ["\\operatorname{conv}(S)", "\\sum f_i", "\\varepsilon\\text{-net}"],
    "concept_count": len(concepts),
    "vandermonde_factor": str(vandermonde),
    "residual_when_two_parameters_coincide": str(residual_at_repeat),
    "concept_length_entropy": entropy,
    "checks": {
        "dependency_graph_is_weakly_connected": nx.is_weakly_connected(G),
        "vandermonde_vanishes_on_collision": residual_at_repeat == 0,
        "entropy_positive": entropy > 0,
    },
}
invariant_path = save_json(invariant_report, CHECKS / "lattices-and-minkowski-s-theorem-invariants.json")
print(json.dumps(invariant_report, indent=2))
```
