        ## Pitfalls

        A common failure mode in this chapter family is to confuse the representation with the invariant. Coordinates can be stretched, surfaces can be embedded differently, frames can rotate, and forms can be written in many bases. The notebook asks the reader to notice which data changes and which data does not. Another failure mode is to accept a diagram without a measurement. Every diagram in this pass is paired with either a label, a ledger, or a saved check. If a figure cannot explain what should be inspected, it does not belong in the chapter.

        Exercise chapters are handled as guided laboratories rather than copied problem lists. The exercises are paraphrased into problem families: construct a model, measure an invariant, compare with a theorem, then perturb the setup. This preserves the mathematical purpose of the exercises without reproducing long source text. It also makes the chapters useful for review because the reader can rerun each lab and adjust the inputs.

        ## Applied Lab

        The lab for this chapter is to rerun the visual sequence after changing one parameter: a radius, a genus, a field winding number, a metric scale, a frame sample, or a path. The expected behavior is stated near each artifact. If the behavior changes when it should not, the model is exposing a misunderstanding. If the behavior remains stable under the right perturbations, the reader has found the invariant that the chapter is trying to teach.

        A useful way to read the lab is to keep a small audit table beside the notebook. One column names the object being manipulated, one column names the measurement, and one column records the invariant that should survive. For example, a curve experiment records a parameterization, an arc-length or turning measurement, and the curvature quantity that should not depend on cosmetic sampling choices. A metric experiment records a chart, a local scale or angle measurement, and the geometric statement that should survive a coordinate change. A forms experiment records the directed input geometry, the form evaluation, and the orientation convention that makes the sign meaningful.

        This audit habit is deliberately repetitive. Differential geometry becomes much less mysterious when the reader can say which part of a computation is a representation choice and which part is geometric evidence. The notebook therefore favors several smaller artifacts over one large summary image. Smaller artifacts make it easier to inspect assumptions, rerun a single construction, and replace a simplified model with a richer one later without disturbing the rest of the chapter.
