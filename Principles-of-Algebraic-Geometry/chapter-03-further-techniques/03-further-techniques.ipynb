## Chapter Goal

This chapter extends the toolkit from smooth forms and bundles to the analytic devices that let algebraic geometry compute with singular or concentrated data. Currents let forms be tested against generalized geometric objects. Residue formulas extract local contributions. Chern-Weil theory turns curvature into characteristic classes. Fixed-point and residue formulas express global invariants as sums of local terms. Spectral sequences organize filtered and bigraded complexes so that complicated cohomology can be computed page by page.

The notebook keeps these techniques connected by one question: where does a global invariant hide when the geometry is local, singular, filtered, or equivariant? A current records how a geometric subset acts on test forms. A smoothing kernel shows how singular data can be approximated without losing its mass. A Chern form is a curvature ledger whose integral becomes a topological number. A fixed-point formula is a localization ledger: each fixed point contributes a rational weight, and the sum is the global trace. A spectral sequence is a moving table whose entries survive, die, or migrate according to differentials.

## Translation Guide

- A current is represented by a functional on test functions and by a concentrated approximation.
- A residue is represented as a local coefficient or contour contribution that should remain stable under allowed transformations.
- Intersections of analytic varieties become local contribution ledgers, not merely pictures of crossing curves.
- Chern classes are represented by curvature forms, traces, and integrals that can be checked on small models.
- Fixed-point and Bott-style formulas become weighted sums over isolated local data.
- Spectral sequences are represented as page tables with differentials, survivors, and convergence targets.

## Currents as Tested Geometry

A current is best understood operationally: it eats a test form and returns a number. The model below approximates a point-supported current by normalized Gaussian kernels of decreasing width. The visual makes two checks visible. The total mass should stay one, and the tested value should converge to the value of the test function at the support. This is the computational analogue of using smoothing and regularity to move between singular geometric data and ordinary differential forms.

```python

x=np.linspace(-3,3,1600)
test=np.cos(1.7*x)*np.exp(-0.15*x*x)+0.4
sigmas=[0.9,0.55,0.32,0.18]
fig,axes=plt.subplots(1,2,figsize=(12,5))
axes[0].plot(x,test,color='black',lw=2,label='test function')
convergence=[]
for sigma in sigmas:
    kernel=np.exp(-x*x/(2*sigma*sigma))/(np.sqrt(2*np.pi)*sigma)
    mass=float(np.trapz(kernel,x))
    value=float(np.trapz(kernel*test,x))
    convergence.append({'sigma':sigma,'mass':mass,'test_value':value})
    axes[0].plot(x,kernel/kernel.max(),label=f'sigma={sigma}')
axes[0].set_title('Smoothing a point current'); axes[0].grid(alpha=.25); axes[0].legend(fontsize=8)
axes[1].plot([c['sigma'] for c in convergence],[c['test_value'] for c in convergence],marker='o')
axes[1].axhline(float(test[len(test)//2]),color='#d62728',ls='--',label='test value at support')
axes[1].invert_xaxis(); axes[1].set_xlabel('kernel width sigma'); axes[1].set_ylabel('tested value'); axes[1].set_title('Regularization converges as support concentrates'); axes[1].grid(alpha=.25); axes[1].legend()
path=savefig(fig,FIG_DIR/'current-smoothing-kernel-convergence.png')
current_checks={'mass_errors':[abs(c['mass']-1.0) for c in convergence],'last_test_error':float(abs(convergence[-1]['test_value']-test[len(test)//2])),'convergence':convergence}
save_json(current_checks,CHECK_DIR/'current-smoothing-kernel-convergence.json')
display_artifact(path)
current_checks
```

## Residues, Intersections, and Local Contributions

Residue formulas and intersection currents share a structural idea: a global object can be reconstructed from local contributions when the hypotheses are right. The next cell computes a residue coefficient exactly and places it beside a finite intersection ledger for two local curves. The ledger is intentionally finite because the learner should inspect which local terms contribute and why their sum is the invariant being used later in residue and fixed-point formulas.

```python

z=sp.symbols('z')
expr=(2+3*z+z**2)/z
residue=sp.residue(expr,z,0)
points=['p0','p1','p2']; local_intersections=[2,1,3]
fig,axes=plt.subplots(1,2,figsize=(11,4.8))
axes[0].bar(['Res at 0'],[float(residue)],color='#f58518'); axes[0].set_ylim(0,3.5); axes[0].set_title('Local residue coefficient')
axes[1].bar(points,local_intersections,color='#54a24b'); axes[1].set_title('Intersection number as local sum'); axes[1].set_ylabel('multiplicity')
for ax in axes: ax.grid(alpha=.2,axis='y')
path=savefig(fig,FIG_DIR/'residue-intersection-localization-ledger.png')
residue_checks={'function':'(2+3z+z^2)/z','residue_at_zero':int(residue),'local_intersections':dict(zip(points,local_intersections)),'intersection_total':int(sum(local_intersections))}
save_json(residue_checks,CHECK_DIR/'residue-intersection-localization-ledger.json')
display_artifact(path)
residue_checks
```

## Chern-Weil and Fixed-Point Formula Dependencies

The chapter's Chern class and fixed-point sections are both localization stories with different inputs. Chern-Weil starts with curvature and produces characteristic forms. Lefschetz and Bott-style formulas start with fixed points, vector-field zeros, or local denominators and produce global traces or characteristic numbers. The graph below is a proof scaffold: it records the flow of data so that the reader can see where local analytic or equivariant information enters the final formula.

```python

G=nx.DiGraph()
G.add_edges_from([
    ('connection','curvature F'),('curvature F','trace polynomial'),('trace polynomial','Chern class'),('Chern class','Gauss-Bonnet / HRR term'),
    ('holomorphic map','isolated fixed points'),('linearized action','local denominator'),('isolated fixed points','Lefschetz sum'),('local denominator','Lefschetz sum'),('Lefschetz sum','global trace'),
    ('vector field zero','Bott residue weight'),('Bott residue weight','characteristic number'),('Chern class','characteristic number'),
    ('filtered complex','spectral sequence'),('spectral sequence','graded cohomology target')])
pos=nx.spring_layout(G,seed=11,k=.65)
fig,ax=plt.subplots(figsize=(11,7.5))
nx.draw_networkx(G,pos,ax=ax,node_color='#e6f4ea',edge_color='#555',node_size=1600,font_size=8,arrows=True,arrowstyle='-|>',arrowsize=13)
ax.set_title('Chern, fixed-point, Bott, and spectral-sequence dependency map'); ax.axis('off')
path=savefig(fig,FIG_DIR/'chern-fixed-point-dependency-map.png')
dep_checks={'nodes':G.number_of_nodes(),'edges':G.number_of_edges(),'global_nodes_have_inputs':all(G.in_degree(n)>0 for n in ['Gauss-Bonnet / HRR term','global trace','characteristic number','graded cohomology target'])}
save_json(dep_checks,CHECK_DIR/'chern-fixed-point-dependency-map.json')
display_artifact(path)
dep_checks
```

## Spectral Sequences as Page Ledgers

A spectral sequence can feel abstract because it is a controlled way of losing and retaining information across pages. This notebook treats each page as a table. Differentials remove paired rank, survivors move forward, and the total visible rank should not increase as the sequence approaches its target. The PNG gives a compact comparison of three pages; the HTML artifact lets the reader inspect each page as a separate heatmap.

```python

E1=np.array([[3,2,0,0],[1,3,2,0],[0,1,2,1],[0,0,1,1]])
E2=np.array([[3,1,0,0],[1,2,1,0],[0,1,1,1],[0,0,1,1]])
Einf=np.array([[2,1,0,0],[1,2,1,0],[0,0,1,1],[0,0,0,1]])
pages=[('E1',E1),('E2',E2),('E∞',Einf)]
fig,axes=plt.subplots(1,3,figsize=(13,4.5))
for ax,(name,arr) in zip(axes,pages):
    im=ax.imshow(arr,cmap='magma',vmin=0,vmax=3)
    for i in range(arr.shape[0]):
        for j in range(arr.shape[1]): ax.text(j,i,str(int(arr[i,j])),ha='center',va='center',color='white' if arr[i,j]>1 else 'black')
    ax.set_title(name); ax.set_xlabel('p'); ax.set_ylabel('q')
fig.colorbar(im,ax=axes.ravel().tolist(),fraction=.025)
path=savefig(fig,FIG_DIR/'spectral-sequence-page-rank-ledger.png')
fig_html=go.Figure()
for name,arr in pages:
    fig_html.add_trace(go.Heatmap(z=arr, visible=(name=='E1'), colorscale='Magma', zmin=0, zmax=3, name=name, showscale=True))
buttons=[]
for k,(name,arr) in enumerate(pages):
    buttons.append(dict(label=name, method='update', args=[{'visible':[i==k for i in range(len(pages))]}, {'title':f'Spectral sequence page {name}'}]))
fig_html.update_layout(title='Spectral sequence page E1', updatemenus=[dict(buttons=buttons, direction='right', x=.1, y=1.16)], width=720, height=520)
html_path=HTML_DIR/'spectral-sequence-page-explorer.html'; fig_html.write_html(html_path,include_plotlyjs='cdn')
spectral_checks={'page_ranks':{name:int(arr.sum()) for name,arr in pages},'rank_never_increases':bool(E1.sum()>=E2.sum()>=Einf.sum()),'target_rank':int(Einf.sum())}
save_json(spectral_checks,CHECK_DIR/'spectral-sequence-page-rank-ledger.json')
display_artifact(path); display_artifact(html_path)
spectral_checks
```

## Takeaways

- Currents let geometry act on test forms; smoothing is a way to inspect singular data without pretending it is smooth.

- Chern classes sit between curvature computations and topological characteristic numbers.
- Spectral sequences are not decoration: they are page-by-page ledgers for filtered or bigraded cohomological information.
