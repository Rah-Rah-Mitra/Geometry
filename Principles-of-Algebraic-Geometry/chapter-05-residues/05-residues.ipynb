## Chapter Goal

This chapter turns residues into a working computational language for algebraic geometry. A residue begins as a local coefficient, but the chapter develops it into a bridge between contour-style local calculations, intersection multiplicities, finite holomorphic mappings, plane projective geometry, commutative algebra, Koszul complexes, coherent sheaves, global Ext, duality, superabundance, vector-bundle extensions, and residues with vector bundles. The notebook teaches that bridge by making the local and global ledgers visible.

A reader should leave with three habits. First, local residues can be computed as coefficients or as contour integrals, and the value should be invariant under allowed coordinate transformations when the Jacobian is included correctly. Second, global residue statements are balance laws: local contributions sum to a constrained global value. Third, homological algebra is not a separate chapter bolted on at the end; the Koszul complex, Ext groups, and duality pairings explain why residue pairings are nondegenerate and why point configurations can have superabundance.

## Translation Guide

- A local residue is represented by an exact coefficient and by a sampled contour integral.
- The global residue theorem is represented by a finite pole ledger whose total satisfies the global balance condition.
- The transformation law is represented by a Jacobian-corrected substitution check.
- Local duality is represented by a pairing matrix between quotient monomials and residue functionals.
- A Koszul complex is represented by explicit matrices with `d1*d2 = 0`.
- Coherent sheaves, global Ext, vector fields with isolated zeros, and vector-bundle extensions are represented by dependency graphs and finite dimension ledgers.

## Local and Global Residues as Balance Laws

A local residue is the coefficient that survives a pole. A global residue theorem says that a collection of local contributions is constrained by the global geometry. The example below uses the rational form `dz/(z(z-1))` on the sphere: the residues at `0` and `1` cancel, and the contribution at infinity is zero. The PNG records the contour picture; the HTML artifact lets the learner inspect the pole ledger as a global balance rather than a list of unrelated coefficients.

```python

z=sp.symbols('z')
form=1/(z*(z-1))
res0=sp.residue(form,z,0); res1=sp.residue(form,z,1); resinf=-(res0+res1)
residues={'0':int(res0),'1':int(res1),'infinity':int(resinf)}
theta=np.linspace(0,2*np.pi,500)
fig,axes=plt.subplots(1,2,figsize=(12,5))
axes[0].plot(np.cos(theta),np.sin(theta),color='#4c78a8',label='contour around 0')
axes[0].scatter([0,1],[0,0],s=90,color=['#d62728','#54a24b']); axes[0].text(0,.12,'0'); axes[0].text(1,.12,'1')
axes[0].set_aspect('equal'); axes[0].grid(alpha=.25); axes[0].set_title('Pole positions and contour sample')
axes[1].bar(list(residues.keys()),list(residues.values()),color=['#d62728','#54a24b','#888'])
axes[1].axhline(0,color='0.3',lw=1); axes[1].set_title('Global residue ledger')
path=savefig(fig,FIG_DIR/'local-global-residue-balance.png')
fig_html=go.Figure(data=[go.Bar(x=list(residues.keys()),y=list(residues.values()),text=[str(v) for v in residues.values()],textposition='auto')])
fig_html.update_layout(title='Global residue pole explorer: contributions sum to zero',yaxis_title='residue')
html_path=HTML_DIR/'global-residue-pole-explorer.html'; fig_html.write_html(html_path,include_plotlyjs='cdn')
residue_checks={'form':'dz/(z(z-1))','residues':residues,'sum_residues':int(sum(residues.values())),'global_balance_zero':sum(residues.values())==0}
save_json(residue_checks,CHECK_DIR/'local-global-residue-balance.json')
display_artifact(path); display_artifact(html_path)
residue_checks
```

## Transformation Law and Local Duality Pairing

Residue transformations are easy to get wrong if the Jacobian is invisible. The first panel records a substitution check: the pole coordinate changes, but the Jacobian-corrected residue remains the same. The second panel records a toy local-duality pairing on a quotient basis. The full theory is much richer, but the finite matrix makes the structural point: residues become pairings, and nondegeneracy is a rank statement.

```python

u=sp.symbols('u')
change=2*u
original=sp.residue(1/z,z,0)
transformed=sp.residue((1/change)*sp.diff(change,u),u,0)
pairing=np.array([[0,1],[1,0]],dtype=int)
fig,axes=plt.subplots(1,2,figsize=(11,4.8))
axes[0].bar(['original Res dz/z','after z=2u'],[float(original),float(transformed)],color=['#4c78a8','#f58518'])
axes[0].set_ylim(0,1.25); axes[0].set_title('Jacobian-corrected transformation law')
im=axes[1].imshow(pairing,cmap='Blues',vmin=0,vmax=1)
axes[1].set_xticks([0,1],['1','x']); axes[1].set_yticks([0,1],['dual 1','dual x']); axes[1].set_title('Local duality pairing matrix')
for i in range(2):
    for j in range(2): axes[1].text(j,i,str(pairing[i,j]),ha='center',va='center')
path=savefig(fig,FIG_DIR/'transformation-local-duality-pairing.png')
duality_checks={'original_residue':int(original),'transformed_residue':int(transformed),'transformation_preserves_residue':bool(original==transformed),'pairing_rank':int(np.linalg.matrix_rank(pairing)),'pairing_full_rank':bool(np.linalg.matrix_rank(pairing)==2)}
save_json(duality_checks,CHECK_DIR/'transformation-local-duality-pairing.json')
display_artifact(path)
duality_checks
```

## Koszul Complex as an Executable Boundary Check

The commutative and homological algebra section prepares the global duality material. A Koszul complex is a concrete place to see homological algebra working: maps are matrices, and being a complex means consecutive maps compose to zero. Here `d2 = [-y, x]^T` and `d1 = [x, y]`, so `d1*d2 = -xy + yx = 0`. The graph names the modules and the JSON stores the exact symbolic check.

```python

x,y=sp.symbols('x y')
d1=sp.Matrix([[x,y]])
d2=sp.Matrix([[-y],[x]])
composition=sp.simplify(d1*d2)
G=nx.DiGraph(); G.add_edges_from([('R','R^2'),('R^2','R'),('R','R/(x,y)')])
pos={'R':(-1,0),'R^2':(0,0),'R':(-1,0),'R/(x,y)':(1,0)}
# Use distinct labels in a parallel drawing graph.
H=nx.DiGraph(); H.add_edge('K2 = R','K1 = R^2',label='d2'); H.add_edge('K1 = R^2','K0 = R',label='d1'); H.add_edge('K0 = R','quotient',label='mod (x,y)')
pos=nx.spring_layout(H,seed=5)
fig,ax=plt.subplots(figsize=(8,4.8)); nx.draw_networkx(H,pos,ax=ax,node_color='#fff3cd',edge_color='#555',node_size=1700,font_size=9,arrows=True,arrowstyle='-|>',arrowsize=14); nx.draw_networkx_edge_labels(H,pos,nx.get_edge_attributes(H,'label'),ax=ax,font_size=8); ax.set_title('Koszul complex boundary check'); ax.axis('off')
path=savefig(fig,FIG_DIR/'koszul-complex-boundary-check.png')
koszul_checks={'d1':[str(v) for v in d1],'d2':[str(v) for v in d2],'composition':str(composition[0]),'boundary_squared_zero':bool(composition[0]==0)}
save_json(koszul_checks,CHECK_DIR/'koszul-complex-boundary-check.json')
display_artifact(path)
koszul_checks
```

## Global Duality, Superabundance, and Vector Bundles

The final section cluster connects residues to global Ext, vector fields with isolated zeros, superabundance of points on a surface, module extensions, rank-two vector bundles, and residues with vector bundles. The dependency graph below is a review map. The companion dimension ledger marks the basic phenomenon: expected conditions can fail to be independent, and the excess is superabundance. This gives the reader a computational hook for a sophisticated global-duality discussion.

```python

G=nx.DiGraph(); G.add_edges_from([('coherent sheaves','global Ext'),('global Ext','global duality'),('residue pairing','global duality'),('vector fields with isolated zeros','Ext residue formula'),('global duality','superabundance ledger'),('superabundance ledger','rank-two vector bundle extension'),('module extensions','rank-two vector bundle extension'),('rank-two vector bundle extension','residues with vector bundles')])
pos=nx.spring_layout(G,seed=31,k=.7)
expected=np.array([5,6,7,8]); actual=np.array([5,5,6,6]); excess=expected-actual
fig,axes=plt.subplots(1,2,figsize=(13,5.5))
nx.draw_networkx(G,pos,ax=axes[0],node_color='#e8f3ff',edge_color='#555',node_size=1600,font_size=8,arrows=True,arrowstyle='-|>',arrowsize=13); axes[0].set_title('Global duality and vector-bundle dependency map'); axes[0].axis('off')
axes[1].plot(expected,actual,marker='o',label='actual independent conditions'); axes[1].plot(expected,expected,ls='--',color='0.4',label='expected line')
for e,a,xs in zip(expected,actual,excess): axes[1].annotate(f'excess {xs}',(e,a),textcoords='offset points',xytext=(5,-12),fontsize=8)
axes[1].set_xlabel('expected conditions'); axes[1].set_ylabel('actual rank'); axes[1].set_title('Superabundance as excess dimension'); axes[1].grid(alpha=.25); axes[1].legend()
path=savefig(fig,FIG_DIR/'global-duality-vector-bundle-ledger.png')
global_checks={'expected_conditions':expected.tolist(),'actual_rank':actual.tolist(),'superabundance':excess.tolist(),'nonnegative_excess':bool(np.all(excess>=0)),'dependency_edges':G.number_of_edges()}
save_json(global_checks,CHECK_DIR/'global-duality-vector-bundle-ledger.json')
display_artifact(path)
global_checks
```
