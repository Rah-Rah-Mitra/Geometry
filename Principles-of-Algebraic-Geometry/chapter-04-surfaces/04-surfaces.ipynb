## Chapter Goal

This chapter is the geometric heart of the book. It studies algebraic surfaces through intersection numbers, adjunction, blowups and blowdowns, quadric and cubic surfaces, rational and birational maps, rational ruled surfaces, surfaces of minimal degree, elliptic and irrational ruled surfaces, K3 and Enriques surfaces, Noether's formula, and singularity corrections. The notebook cannot replace the full classification arguments, but it can make the recurring data visible: divisors become lattice vectors, blowups alter intersection forms, birational maps become resolution graphs, rulings become families of curves, and classification types occupy regions in an invariant geography.

The most important computational habit is to keep ledgers. A surface calculation rarely uses a single number. It uses an intersection matrix, a canonical class, Euler characteristic, arithmetic genus, self-intersection, exceptional curves, singularity corrections, and fibration data. The code below turns those ledgers into inspectable artifacts. The reader should be able to rerun the notebook, modify a divisor vector or a blowup count, and see how adjunction, Noether-style identities, and classification summaries change.

## Translation Guide

- Intersection numbers are matrix products on divisor classes.
- Adjunction is a check tying a curve's self-intersection, its canonical intersection, and its genus.
- A blowup subtracts an exceptional direction and changes the intersection form by adding a `-1` class.
- Quadric surfaces are read through two visible rulings; cubic surfaces are represented through incidence and exceptional-line bookkeeping.
- Birational maps between surfaces are represented by resolution graphs rather than by a single rational formula.
- Rational, ruled, elliptic, K3, and Enriques surfaces are compared through invariants such as `K^2`, `chi`, `q`, `p_g`, and fibration data.
- Noether's formula and singularity corrections become finite arithmetic ledgers.

## Intersection Theory and Adjunction as Matrix Arithmetic

A surface turns divisor classes into an intersection lattice. Once a basis is chosen, intersection numbers become matrix products. Adjunction then ties a curve class to its genus through its self-intersection and canonical intersection. This is the right computational starting point because the rest of the chapter repeatedly changes the lattice by blowups, ruled-surface generators, exceptional curves, and singularity corrections.

```python

# Basis H, E1, E2 for the blowup of the projective plane at two points.
Q=sp.diag(1,-1,-1)
K=sp.Matrix([-3,1,1])
classes={'line H':sp.Matrix([1,0,0]),'conic 2H-E1-E2':sp.Matrix([2,-1,-1]),'exceptional E1':sp.Matrix([0,1,0]),'cubic 3H-E1-E2':sp.Matrix([3,-1,-1])}
ledger=[]
for name,C in classes.items():
    self_int=int((C.T*Q*C)[0])
    Kdot=int((K.T*Q*C)[0])
    genus=int((self_int+Kdot)//2 + 1)
    ledger.append({'class':name,'self_intersection':self_int,'K_dot_C':Kdot,'adjunction_genus':genus,'adjunction_left':2*genus-2,'adjunction_right':self_int+Kdot})
fig,axes=plt.subplots(1,2,figsize=(12,5))
im=axes[0].imshow(np.array(Q.tolist(),dtype=float),cmap='coolwarm',vmin=-1,vmax=1)
axes[0].set_xticks(range(3),['H','E1','E2']); axes[0].set_yticks(range(3),['H','E1','E2']); axes[0].set_title('Intersection form after two blowups')
for i in range(3):
    for j in range(3): axes[0].text(j,i,str(Q[i,j]),ha='center',va='center')
axes[1].bar([r['class'] for r in ledger],[r['adjunction_genus'] for r in ledger],color='#4c78a8')
axes[1].set_xticklabels([r['class'] for r in ledger],rotation=25,ha='right'); axes[1].set_title('Adjunction genus ledger'); axes[1].set_ylabel('genus')
path=savefig(fig,FIG_DIR/'intersection-adjunction-ledger.png')
checks={'intersection_matrix':np.array(Q.tolist(),dtype=int).tolist(),'canonical_class':list(map(int,K)),'ledger':ledger,'adjunction_balances':all(r['adjunction_left']==r['adjunction_right'] for r in ledger)}
save_json(checks,CHECK_DIR/'intersection-adjunction-ledger.json')
display_artifact(path)
checks
```

## Blowups, Blowdowns, and Birational Resolution Graphs

A birational map can be indeterminate at points, and the surface chapter repeatedly resolves such maps by blowing up those points. The graph below records a small resolution pattern: the original surface maps rationally to a target, a resolved surface sits above it, and exceptional curves record where points were replaced by directions. The negative diagonal entries in the transformed intersection form are the arithmetic shadow of those exceptional curves.

```python

G=nx.DiGraph()
G.add_edges_from([('S','rational map to T'),('resolved surface S~','S'),('resolved surface S~','T'),('blowup p1','resolved surface S~'),('blowup p2','resolved surface S~'),('exceptional E1','blowup p1'),('exceptional E2','blowup p2'),('contract -1 curve','minimal model')])
pos=nx.spring_layout(G,seed=19,k=.8)
fig,axes=plt.subplots(1,2,figsize=(13,5.7))
nx.draw_networkx(G,pos,ax=axes[0],node_color='#fcefe3',edge_color='#555',node_size=1500,font_size=8,arrows=True,arrowstyle='-|>',arrowsize=13)
axes[0].set_title('Birational map resolved by blowups'); axes[0].axis('off')
exceptional_form=np.diag([1,-1,-1,-1])
im=axes[1].imshow(exceptional_form,cmap='coolwarm',vmin=-1,vmax=1)
labels=['H','E1','E2','E3']; axes[1].set_xticks(range(4),labels); axes[1].set_yticks(range(4),labels)
for i in range(4):
    for j in range(4): axes[1].text(j,i,str(exceptional_form[i,j]),ha='center',va='center')
axes[1].set_title('Intersection form after three blowups')
path=savefig(fig,FIG_DIR/'blowup-birational-resolution-graph.png')
blowup_checks={'exceptional_diagonal':exceptional_form.diagonal().astype(int).tolist(),'negative_exceptional_count':int(np.sum(exceptional_form.diagonal()<0)),'resolution_graph_edges':G.number_of_edges()}
save_json(blowup_checks,CHECK_DIR/'blowup-birational-resolution-graph.json')
display_artifact(path)
blowup_checks
```

## Quadric Rulings and Cubic-Surface Incidence

The quadric surface is the first surface in the chapter where a visual model carries real information: it has two distinct rulings. The Plotly artifact samples the standard parametrization of a smooth quadric and overlays curves from both families. The companion check verifies the affine quadric relation on the sampled grid. The same mindset extends to cubic surfaces, where incidence among lines becomes a combinatorial object with geometric meaning.

```python

u=np.linspace(-1.4,1.4,45); v=np.linspace(-1.4,1.4,45)
U,V=np.meshgrid(u,v)
X=U; Y=V; Z=U*V
fig=go.Figure(data=[go.Surface(x=X,y=Y,z=Z,opacity=.65,colorscale='Viridis',showscale=False,name='quadric chart')])
for c in [-1,-.35,.35,1]:
    uu=np.linspace(-1.4,1.4,100); vv=np.full_like(uu,c)
    fig.add_trace(go.Scatter3d(x=uu,y=vv,z=uu*vv,mode='lines',line=dict(width=5,color='crimson'),name=f'v={c} ruling'))
    vv2=np.linspace(-1.4,1.4,100); uu2=np.full_like(vv2,c)
    fig.add_trace(go.Scatter3d(x=uu2,y=vv2,z=uu2*vv2,mode='lines',line=dict(width=5,color='royalblue'),name=f'u={c} ruling'))
fig.update_layout(title='Quadric surface chart with two ruling families',width=780,height=620,scene=dict(xaxis_title='u',yaxis_title='v',zaxis_title='uv'))
html_path=HTML_DIR/'quadric-surface-rulings.html'; fig.write_html(html_path,include_plotlyjs='cdn')
H=nx.Graph(); H.add_edges_from([('L1','M1'),('L1','M2'),('L2','M1'),('L2','M3'),('L3','M2'),('L3','M3'),('E1','L1'),('E2','L2'),('E3','L3')])
pos=nx.spring_layout(H,seed=23)
fig2,ax=plt.subplots(figsize=(7,5.5)); nx.draw_networkx(H,pos,ax=ax,node_color='#e8f3ff',edge_color='#555',node_size=1100,font_size=8); ax.set_title('Sample cubic-surface incidence scaffold'); ax.axis('off')
inc_path=savefig(fig2,FIG_DIR/'quadric-cubic-incidence-scaffold.png')
quadric_residual=np.max(np.abs(Z-U*V))
quadric_checks={'quadric_chart_equation':'z = u*v','max_residual':float(quadric_residual),'sampled_ruling_count':8,'incidence_edges':H.number_of_edges()}
save_json(quadric_checks,CHECK_DIR/'quadric-cubic-incidence-scaffold.json')
display_artifact(html_path); display_artifact(inc_path)
quadric_checks
```
