## Branched Coverings, Riemann-Hurwitz, And Plane-Curve Genus

A branched covering can be reduced to a ledger: source genus, target genus, degree, and the ramification excess at each branch point. The Riemann-Hurwitz formula says the ledger closes only when the ramification term supplies the missing Euler characteristic. For smooth plane curves, adjunction turns the degree into the genus `(d-1)(d-2)/2`; the same degree also feeds the smooth plane Plucker counts for the dual class and flexes.


````python
def riemann_hurwitz_residual(degree, base_genus, source_genus, ramification_excesses):
    left = 2 * source_genus - 2
    right = degree * (2 * base_genus - 2) + sum(ramification_excesses)
    return left - right

rh_examples = [
    {'case': 'power map z -> z^3 on P1', 'degree': 3, 'base_genus': 0, 'source_genus': 0, 'ramification_excesses': [2, 2]},
    {'case': 'hyperelliptic genus 3 double cover', 'degree': 2, 'base_genus': 0, 'source_genus': 3, 'ramification_excesses': [1] * 8},
    {'case': 'unramified degree 2 elliptic cover', 'degree': 2, 'base_genus': 1, 'source_genus': 1, 'ramification_excesses': []},
]
for row in rh_examples:
    row['ramification_sum'] = sum(row['ramification_excesses'])
    row['residual'] = riemann_hurwitz_residual(row['degree'], row['base_genus'], row['source_genus'], row['ramification_excesses'])

fig, ax = plt.subplots(figsize=(10, 4.8))
base_x = np.linspace(-3.2, 3.2, 500)
branch_x = np.linspace(-2.8, 2.8, 8)
y_top = 1.25 + 0.22 * np.sin(2.2 * base_x)
y_bottom = -1.25 - 0.22 * np.sin(2.2 * base_x)
ax.plot(base_x, y_top, color='#1f6f8b', lw=2.4, label='sheet +')
ax.plot(base_x, y_bottom, color='#1f6f8b', lw=2.4, label='sheet -')
ax.plot(base_x, np.zeros_like(base_x), color='#3b3b3b', lw=2.0, label='base P1 coordinate')
for bx in branch_x:
    ax.plot([bx, bx], [-1.1, 1.1], color='#b44b39', lw=1.2, alpha=0.75)
    ax.scatter([bx], [0], s=34, color='#b44b39', zorder=4)
    ax.text(bx, 0.16, 'e-1=1', ha='center', va='bottom', fontsize=8, rotation=90)
ax.text(-3.15, 1.62, 'source curve: two local sheets', fontsize=11, weight='bold')
ax.text(-3.15, -0.42, 'target P1', fontsize=10)
ax.text(-3.15, -1.9, '2g-2 = 2(-2) + (2g+2); for g=3: 4 = -4 + 8', fontsize=11)
ax.set_title('Riemann-Hurwitz as a branch ledger')
ax.set_xlim(-3.4, 3.4)
ax.set_ylim(-2.2, 2.1)
ax.axis('off')
fig.tight_layout()
branch_fig = FIG_DIR / 'branched-covering-riemann-hurwitz-ledger.svg'
fig.savefig(branch_fig, format='svg')
plt.close(fig)

d = sp.symbols('d', integer=True, positive=True)
genus_formula = sp.simplify((d - 1) * (d - 2) / 2)
degrees = list(range(1, 9))
genus_rows = []
for degree in degrees:
    genus = int(genus_formula.subs(d, degree))
    dual_class = degree * (degree - 1)
    flex_count = 3 * degree * (degree - 2) if degree >= 2 else 0
    genus_rows.append({
        'degree': degree,
        'smooth_plane_genus': genus,
        'smooth_dual_class': dual_class,
        'smooth_flex_count': flex_count,
        'genus_residual': genus - ((degree - 1) * (degree - 2) // 2),
    })

fig, axes = plt.subplots(1, 2, figsize=(11, 4.4))
axes[0].plot(degrees, [r['smooth_plane_genus'] for r in genus_rows], marker='o', color='#365f91')
axes[0].set_title('Smooth plane curve genus')
axes[0].set_xlabel('degree d')
axes[0].set_ylabel('g = (d-1)(d-2)/2')
axes[0].grid(alpha=0.25)
width = 0.36
x = np.arange(len(degrees))
axes[1].bar(x - width/2, [r['smooth_dual_class'] for r in genus_rows], width, label='dual class d(d-1)', color='#6b8e23')
axes[1].bar(x + width/2, [r['smooth_flex_count'] for r in genus_rows], width, label='flex count 3d(d-2)', color='#c26a2e')
axes[1].set_xticks(x)
axes[1].set_xticklabels(degrees)
axes[1].set_xlabel('degree d')
axes[1].set_title('Smooth plane Plucker rows')
axes[1].legend(fontsize=9)
axes[1].grid(axis='y', alpha=0.25)
fig.tight_layout()
genus_fig = FIG_DIR / 'plane-curve-genus-and-plucker-ledger.svg'
fig.savefig(genus_fig, format='svg')
plt.close(fig)

write_json(CHECK_DIR / 'riemann-hurwitz-ledger.json', {'formula': '2g(S)-2 = n(2g(base)-2) + sum(e_p-1)', 'examples': rh_examples})
write_json(CHECK_DIR / 'genus-formula-checks.json', {'formula': 'g=(d-1)(d-2)/2 for smooth plane curves', 'rows': genus_rows})
write_csv(TABLE_DIR / 'plane-curve-genus-plucker-ledger.csv', genus_rows,
          ['degree', 'smooth_plane_genus', 'smooth_dual_class', 'smooth_flex_count', 'genus_residual'])

assert all(row['residual'] == 0 for row in rh_examples)
assert all(row['genus_residual'] == 0 for row in genus_rows)

display_artifact('artifacts/chapter-02/figures/branched-covering-riemann-hurwitz-ledger.svg')
display_artifact('artifacts/chapter-02/figures/plane-curve-genus-and-plucker-ledger.svg')
````

## Abel Sums And The Jacobian Torus

The chapter's Abel theorem turns a divisor of degree zero into a point of the Jacobian. Computationally, this is a vector sum in a period lattice. The figure uses a genus-one period parallelogram as a readable model: an unwrapped path in the complex plane is folded back into the quotient, and a triple whose coordinates add to an integral lattice vector represents the zero class.


````python
tau = 0.34 + 1.18j


def lattice_to_complex(x_coord, y_coord):
    return x_coord + y_coord * tau


def wrap_lattice(coords):
    return coords - np.floor(coords)


divisor_lattice_coords = np.array([[0.13, 0.22], [0.41, 0.31], [0.46, 0.47]])
divisor_sum = divisor_lattice_coords.sum(axis=0)
lattice_residual = divisor_sum - np.rint(divisor_sum)

t = np.linspace(0, 1, 240)
unwrapped_lattice = np.column_stack([
    0.08 + 1.28 * t,
    0.16 + 0.92 * t + 0.08 * np.sin(2 * np.pi * t),
])
wrapped_lattice = np.array([wrap_lattice(row) for row in unwrapped_lattice])
unwrapped_z = np.array([lattice_to_complex(x0, y0) for x0, y0 in unwrapped_lattice])
wrapped_z = np.array([lattice_to_complex(x0, y0) for x0, y0 in wrapped_lattice])

fig, axes = plt.subplots(1, 2, figsize=(11, 4.8))
parallelogram = np.array([0, 1, 1 + tau, tau, 0], dtype=complex)
for ax in axes:
    ax.plot(parallelogram.real, parallelogram.imag, color='#333333', lw=1.7)
    ax.set_aspect('equal', adjustable='box')
    ax.set_xlabel('Re')
    ax.set_ylabel('Im')
    ax.grid(alpha=0.2)
axes[0].plot(unwrapped_z.real, unwrapped_z.imag, color='#8a4f9e', lw=2.2)
axes[0].set_title('Abel path before quotient')
axes[1].plot(wrapped_z.real, wrapped_z.imag, color='#0f7b6c', lw=2.2)
axes[1].scatter([lattice_to_complex(x0, y0).real for x0, y0 in divisor_lattice_coords],
                [lattice_to_complex(x0, y0).imag for x0, y0 in divisor_lattice_coords],
                color='#b5483a', s=42, zorder=4, label='triple summands')
axes[1].set_title('Same data modulo periods')
axes[1].legend(fontsize=9)
fig.suptitle('Abel-Jacobian torus path and lattice closure', y=1.02)
fig.tight_layout()
abel_fig = FIG_DIR / 'abel-jacobian-torus-path.png'
fig.savefig(abel_fig, dpi=180, bbox_inches='tight')
plt.close(fig)

plot_fig = go.Figure()
plot_fig.add_trace(go.Scatter(x=wrapped_lattice[:, 0], y=wrapped_lattice[:, 1], mode='lines', name='wrapped Abel path'))
plot_fig.add_trace(go.Scatter(x=divisor_lattice_coords[:, 0], y=divisor_lattice_coords[:, 1], mode='markers+text',
                              text=['p1', 'p2', 'p3'], textposition='top center', name='summands'))
plot_fig.update_layout(
    title='Abel-Jacobian path in fundamental lattice coordinates',
    xaxis_title='period coordinate x', yaxis_title='period coordinate y',
    xaxis=dict(range=[0, 1], constrain='domain'), yaxis=dict(range=[0, 1], scaleanchor='x', scaleratio=1),
    width=820, height=620,
)
abel_html = HTML_DIR / 'abel-jacobian-torus-path.html'
plot_fig.write_html(abel_html, include_plotlyjs=True, full_html=True)

abel_checks = {
    'tau': {'real': tau.real, 'imag': tau.imag},
    'divisor_lattice_coords': divisor_lattice_coords.round(6).tolist(),
    'divisor_sum': divisor_sum.round(6).tolist(),
    'nearest_lattice_vector': np.rint(divisor_sum).astype(int).tolist(),
    'lattice_residual_norm': float(np.linalg.norm(lattice_residual)),
}
write_json(CHECK_DIR / 'abel-jacobian-torus-checks.json', abel_checks)
assert abel_checks['lattice_residual_norm'] < 1e-12

display_artifact('artifacts/chapter-02/figures/abel-jacobian-torus-path.png')
display_artifact('artifacts/chapter-02/html/abel-jacobian-torus-path.html')
````

## Linear Systems And Riemann-Roch Dimension Balance

A divisor is a constraint ledger. The Riemann-Roch formula says that the dimension gained by allowing poles along `D` is balanced by the dimension of residual canonical differentials. The plot shows a generic degree sweep, then adds named special rows: the zero divisor, the canonical divisor, and a hyperelliptic pencil.


````python
def rr_residual(genus, degree, h0_d, h0_k_minus_d):
    return h0_d - h0_k_minus_d - (degree + 1 - genus)

rr_genus = 5
canonical_degree = 2 * rr_genus - 2
rr_rows = []
for degree in range(-2, 2 * rr_genus + 5):
    if degree < 0:
        h0_d = 0
        h0_k_minus_d = rr_genus - 1 - degree
        model = 'generic negative degree'
    elif degree <= rr_genus - 1:
        h0_d = 0
        h0_k_minus_d = rr_genus - 1 - degree
        model = 'generic nonspecial degree range'
    else:
        h0_d = degree + 1 - rr_genus
        h0_k_minus_d = 0
        model = 'generic high degree'
    rr_rows.append({
        'model': model,
        'genus': rr_genus,
        'degree': degree,
        'h0_D': h0_d,
        'h0_K_minus_D': h0_k_minus_d,
        'rr_right_side': degree + 1 - rr_genus,
        'rr_residual': rr_residual(rr_genus, degree, h0_d, h0_k_minus_d),
    })

special_rows = [
    {'model': 'trivial divisor D=0', 'genus': rr_genus, 'degree': 0, 'h0_D': 1, 'h0_K_minus_D': rr_genus},
    {'model': 'canonical divisor D=K', 'genus': rr_genus, 'degree': canonical_degree, 'h0_D': rr_genus, 'h0_K_minus_D': 1},
    {'model': 'hyperelliptic pencil g^1_2', 'genus': rr_genus, 'degree': 2, 'h0_D': 2, 'h0_K_minus_D': rr_genus - 1},
]
for row in special_rows:
    row['rr_right_side'] = row['degree'] + 1 - row['genus']
    row['rr_residual'] = rr_residual(row['genus'], row['degree'], row['h0_D'], row['h0_K_minus_D'])

fig, ax = plt.subplots(figsize=(10, 5))
degrees = [row['degree'] for row in rr_rows]
ax.plot(degrees, [row['h0_D'] for row in rr_rows], marker='o', color='#245b8f', label='generic h0(D)')
ax.plot(degrees, [row['h0_K_minus_D'] for row in rr_rows], marker='s', color='#b25538', label='generic h0(K-D)')
ax.axvline(rr_genus - 1, color='#777777', ls='--', lw=1.2, label='degree g-1')
for row, marker, color in zip(special_rows, ['*', 'P', 'X'], ['#222222', '#7a4ca0', '#0c7c59']):
    ax.scatter([row['degree']], [row['h0_D']], s=110, marker=marker, color=color, zorder=5, label=row['model'])
ax.set_xlabel('degree of D')
ax.set_ylabel('dimension')
ax.set_title('Riemann-Roch dimension balance for genus 5')
ax.grid(alpha=0.25)
ax.legend(fontsize=8, ncol=2)
fig.tight_layout()
rr_fig = FIG_DIR / 'riemann-roch-dimension-balance.svg'
fig.savefig(rr_fig, format='svg')
plt.close(fig)

all_rr_rows = rr_rows + special_rows
write_json(CHECK_DIR / 'riemann-roch-dimension-balance.json', {'formula': 'h0(D)-h0(K-D)=deg(D)+1-g', 'rows': all_rr_rows})
write_csv(TABLE_DIR / 'riemann-roch-dimension-balance.csv', all_rr_rows,
          ['model', 'genus', 'degree', 'h0_D', 'h0_K_minus_D', 'rr_right_side', 'rr_residual'])
assert all(row['rr_residual'] == 0 for row in all_rr_rows)

display_artifact('artifacts/chapter-02/figures/riemann-roch-dimension-balance.svg')
display_artifact('artifacts/chapter-02/tables/riemann-roch-dimension-balance.csv')
````

The nonhyperelliptic case behaves like an embedding; the hyperelliptic case factors through a two-to-one map to `P1`, so conjugate points are folded together. ## Plucker Formulas, Ramification, And Weierstrass Weights

For a base-point-free linear system `g^r_d`, ramification is finite and has a total weight. The chapter uses this repeatedly: first for general Plucker bookkeeping, then for canonical systems, Weierstrass points, and plane-curve counts. The ledger keeps the formula visible rather than burying it in prose.


````python
def plucker_total_weight(genus, r, degree):
    return (r + 1) * (degree + r * (genus - 1))

plucker_cases = [
    {'case': 'hyperelliptic pencil g^1_2, g=3', 'genus': 3, 'r': 1, 'degree': 2},
    {'case': 'trigonal-style g^1_3, g=4', 'genus': 4, 'r': 1, 'degree': 3},
    {'case': 'canonical system, g=4', 'genus': 4, 'r': 3, 'degree': 6},
    {'case': 'canonical system, g=5', 'genus': 5, 'r': 4, 'degree': 8},
]
for row in plucker_cases:
    row['total_weight'] = plucker_total_weight(row['genus'], row['r'], row['degree'])
    if row['case'].startswith('canonical'):
        row['canonical_expected_g3_minus_g'] = row['genus'] ** 3 - row['genus']
        row['canonical_residual'] = row['total_weight'] - row['canonical_expected_g3_minus_g']
    else:
        row['canonical_expected_g3_minus_g'] = None
        row['canonical_residual'] = None

plane_plucker_rows = []
for degree in [3, 4, 5, 6]:
    plane_plucker_rows.append({
        'degree': degree,
        'smooth_genus': (degree - 1) * (degree - 2) // 2,
        'smooth_dual_class': degree * (degree - 1),
        'smooth_flex_count': 3 * degree * (degree - 2),
    })

fig, axes = plt.subplots(1, 2, figsize=(12, 4.8))
case_labels = [row['case'].split(',')[0] for row in plucker_cases]
axes[0].bar(np.arange(len(plucker_cases)), [row['total_weight'] for row in plucker_cases], color=['#5181b8', '#5181b8', '#9a5ca8', '#9a5ca8'])
axes[0].set_xticks(np.arange(len(plucker_cases)))
axes[0].set_xticklabels(case_labels, rotation=18, ha='right')
axes[0].set_ylabel('total ramification weight')
axes[0].set_title('General Plucker ramification ledger')
axes[0].grid(axis='y', alpha=0.25)
axes[1].plot([row['degree'] for row in plane_plucker_rows], [row['smooth_dual_class'] for row in plane_plucker_rows], marker='o', label='dual class')
axes[1].plot([row['degree'] for row in plane_plucker_rows], [row['smooth_flex_count'] for row in plane_plucker_rows], marker='s', label='flex count')
axes[1].set_xlabel('smooth plane degree')
axes[1].set_title('Plane Plucker smooth rows')
axes[1].legend()
axes[1].grid(alpha=0.25)
fig.tight_layout()
plucker_fig = FIG_DIR / 'plucker-ramification-ledger.svg'
fig.savefig(plucker_fig, format='svg')
plt.close(fig)

plucker_payload = {
    'general_formula': 'total ramification weight of g^r_d is (r+1)(d+r(g-1))',
    'cases': plucker_cases,
    'smooth_plane_rows': plane_plucker_rows,
}
write_json(CHECK_DIR / 'plucker-ramification-ledger.json', plucker_payload)
write_csv(TABLE_DIR / 'plucker-ramification-ledger.csv', plucker_cases,
          ['case', 'genus', 'r', 'degree', 'total_weight', 'canonical_expected_g3_minus_g', 'canonical_residual'])
assert all(row['canonical_residual'] in (None, 0) for row in plucker_cases)

display_artifact('artifacts/chapter-02/figures/plucker-ramification-ledger.svg')
````

## Complex Tori, Riemann Conditions, And Theta Functions

A complex torus becomes an Abelian variety only when the lattice carries a compatible positive form. In normalized genus-one notation the theta function is a section with automorphy factors. The heatmap samples its magnitude on the fundamental parallelogram; the check tests the two basic quasi-periodicity laws numerically.


````python
theta_tau = 0.72j


def theta_genus_one(z, tau_value=theta_tau, cutoff=24):
    n = np.arange(-cutoff, cutoff + 1)
    return np.sum(np.exp(np.pi * 1j * (n ** 2) * tau_value + 2 * np.pi * 1j * n * z))

xs = np.linspace(0, 1, 150)
ys = np.linspace(0, 1, 150)
heat = np.zeros((len(ys), len(xs)))
for i, y0 in enumerate(ys):
    for j, x0 in enumerate(xs):
        z = x0 + y0 * theta_tau
        heat[i, j] = math.log10(abs(theta_genus_one(z)) + 1e-10)

fig, ax = plt.subplots(figsize=(7.4, 6.2))
mesh = ax.imshow(heat, extent=[0, 1, 0, 1], origin='lower', cmap='magma', aspect='equal')
ax.set_title('Theta-function magnitude on a period lattice')
ax.set_xlabel('period coordinate x')
ax.set_ylabel('period coordinate y')
fig.colorbar(mesh, ax=ax, label='log10 |theta(z)|')
fig.tight_layout()
theta_png = FIG_DIR / 'theta-function-lattice-heatmap.png'
fig.savefig(theta_png, dpi=180)
plt.close(fig)

theta_plot = go.Figure(data=go.Heatmap(z=heat, x=xs, y=ys, colorscale='Magma', colorbar=dict(title='log10 |theta|')))
theta_plot.update_layout(title='Theta heatmap on the fundamental lattice square', xaxis_title='x', yaxis_title='y', width=800, height=720)
theta_html = HTML_DIR / 'theta-function-lattice-heatmap.html'
theta_plot.write_html(theta_html, include_plotlyjs=True, full_html=True)

z0 = 0.23 + 0.19 * theta_tau
period_one_residual = abs(theta_genus_one(z0 + 1) - theta_genus_one(z0))
automorphy = np.exp(-np.pi * 1j * theta_tau - 2 * np.pi * 1j * z0)
period_tau_residual = abs(theta_genus_one(z0 + theta_tau) - automorphy * theta_genus_one(z0))
theta_checks = {
    'tau': {'real': float(theta_tau.real), 'imag': float(theta_tau.imag)},
    'riemann_condition_imaginary_part_positive': bool(theta_tau.imag > 0),
    'theta_z_plus_1_residual': float(period_one_residual),
    'theta_z_plus_tau_residual': float(period_tau_residual),
    'heat_min': float(np.min(heat)),
    'heat_max': float(np.max(heat)),
    'sample_grid_shape': list(heat.shape),
    'riemann_condition_matrix': [[float(theta_tau.imag)]],
}
write_json(CHECK_DIR / 'theta-lattice-checks.json', theta_checks)
assert theta_checks['riemann_condition_imaginary_part_positive']
assert theta_checks['theta_z_plus_1_residual'] < 1e-10
assert theta_checks['theta_z_plus_tau_residual'] < 1e-8
assert theta_checks['heat_max'] > theta_checks['heat_min']

display_artifact('artifacts/chapter-02/figures/theta-function-lattice-heatmap.png')
display_artifact('artifacts/chapter-02/html/theta-function-lattice-heatmap.html')
````

## Jacobians, Theta Divisors, And Torelli Reconstruction

The final part of the chapter turns the Jacobian from a target of Abelian integrals into a reconstruction device. The dependency graph records the proof strategy: periods and the principal polarization define the Jacobian; Riemann's theorem identifies the theta divisor with a translate of `W_{g-1}`; singularities and Gauss-map data recover the canonical curve; Torelli says that this recovers the original curve.


````python
T = nx.DiGraph()
torelli_edges = [
    ('Curve S', 'Holomorphic differentials H0(K)'),
    ('Curve S', 'Integral homology H1(S,Z)'),
    ('Holomorphic differentials H0(K)', 'Period matrix'),
    ('Integral homology H1(S,Z)', 'Period matrix'),
    ('Period matrix', 'Jacobian J(S)'),
    ('Integral homology H1(S,Z)', 'Principal polarization'),
    ('Principal polarization', 'Theta divisor'),
    ('Jacobian J(S)', 'Theta divisor'),
    ('Theta divisor', 'W_{g-1} translate'),
    ('W_{g-1} translate', 'Theta singularities'),
    ('Theta divisor', 'Gauss map branch data'),
    ('Gauss map branch data', 'Dual canonical curve'),
    ('Dual canonical curve', 'Canonical curve'),
    ('Canonical curve', 'Curve S reconstructed'),
    ('Theta singularities', 'Special linear systems'),
]
T.add_edges_from(torelli_edges)
pos = {
    'Curve S': (0, 2),
    'Holomorphic differentials H0(K)': (1.6, 2.9),
    'Integral homology H1(S,Z)': (1.6, 1.2),
    'Period matrix': (3.2, 2.35),
    'Principal polarization': (3.2, 1.15),
    'Jacobian J(S)': (4.7, 2.35),
    'Theta divisor': (6.1, 1.85),
    'W_{g-1} translate': (7.6, 2.55),
    'Theta singularities': (9.1, 2.8),
    'Special linear systems': (10.7, 2.8),
    'Gauss map branch data': (7.6, 1.15),
    'Dual canonical curve': (9.1, 1.15),
    'Canonical curve': (10.7, 1.15),
    'Curve S reconstructed': (12.4, 1.15),
}
fig, ax = plt.subplots(figsize=(14, 5.6))
nx.draw_networkx_nodes(T, pos, node_color='#edf4f2', edgecolors='#253238', linewidths=1.1, node_size=2250, ax=ax)
nx.draw_networkx_labels(T, pos, font_size=8.5, ax=ax)
nx.draw_networkx_edges(T, pos, arrows=True, arrowstyle='-|>', arrowsize=16, width=1.4, edge_color='#52758a', ax=ax)
ax.set_title('Torelli/Jacobian dependency map')
ax.axis('off')
fig.tight_layout()
torelli_fig = FIG_DIR / 'torelli-jacobian-dependency-map.svg'
fig.savefig(torelli_fig, format='svg')
plt.close(fig)

torelli_checks = {
    'nodes': list(T.nodes()),
    'edges': [{'source': u, 'target': v} for u, v in T.edges()],
    'acyclic': nx.is_directed_acyclic_graph(T),
    'required_terminal': 'Curve S reconstructed' in T.nodes(),
    'path_from_curve_to_reconstruction': nx.has_path(T, 'Curve S', 'Curve S reconstructed'),
}
write_json(CHECK_DIR / 'torelli-dependency-checks.json', torelli_checks)
assert torelli_checks['acyclic']
assert torelli_checks['path_from_curve_to_reconstruction']

display_artifact('artifacts/chapter-02/figures/torelli-jacobian-dependency-map.svg')
````
