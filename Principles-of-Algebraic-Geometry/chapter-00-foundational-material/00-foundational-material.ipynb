## Chapter Goal

This chapter is the equipment room for the rest of the course. It prepares the reader to move between local complex analysis, complex manifolds, analytic varieties, sheaves, topology, vector bundles, curvature, harmonic representatives, and Kahler identities. The goal of the notebook is not to compress every proof into a slogan. The goal is to make the objects visible enough that later chapters can use them without asking the reader to keep the entire foundation in short-term memory.

A useful way to read the chapter computationally is to track three recurring moves. First, a local analytic object becomes a model in coordinates: Cauchy integrals recover coefficients, Weierstrass preparation turns local zero sets into finite covers, and tangent cones read the first nonzero homogeneous term. Second, local data is glued: charts, sheaves, Cech cochains, transition functions, and forms all ask whether compatible local records define a global object. Third, analysis turns topology into geometry: de Rham and Dolbeault cohomology, harmonic theory, curvature forms, Hodge decomposition, and Lefschetz operators turn qualitative existence results into inspectable graded ledgers.

## Translation Guide

- A holomorphic function is treated as something we can sample on a contour and verify by an integral residue or coefficient check.
- An analytic variety is represented by equations, singular strata, tangent spaces, and tangent cones rather than by a decorative curve sketch.
- A complex manifold is represented by chart changes and by the split between holomorphic and antiholomorphic directions.
- A sheaf is represented as local records plus restriction maps; cohomology measures the obstruction to solving a gluing problem globally.
- Intersection theory is represented by signed or oriented pairings, with Poincare duality as the bridge from geometry to linear algebra.
- A vector bundle with connection is represented by transition data, curvature, and characteristic-class ledgers.
- Harmonic and Kahler theory is represented by decomposition tables, symmetry checks, and the Lefschetz ladder.

The notebook favors small exact or numerical experiments over heavy formal machinery. This is deliberate: the later algebraic geometry chapters need the reader to recognize which invariant is being invoked, which hypothesis makes it legal, and what local failure would look like.

## Local Holomorphic Control: Cauchy Meets Weierstrass

The chapter begins with several complex variables, but the first habit is already visible in one variable: holomorphic data inside a contour is controlled by boundary values. In higher-dimensional local analytic geometry, Weierstrass preparation gives a complementary picture: near a suitable point, a hypersurface can be read as a finite branched cover over a coordinate polydisc. The figure below puts those two habits together. The contour samples a holomorphic function and checks Cauchy's formula numerically; the branched cover sketches the local model `w = z^2`, where two sheets come together at the branch point. This is the picture the later theory keeps reusing when a variety is studied through projections, ramification, and local equations.

```python

theta = np.linspace(0, 2*np.pi, 1600, endpoint=False)
z = np.exp(1j*theta)
dz = 1j*z*(2*np.pi/len(theta))
f = lambda u: 1/(1 - 0.3*u)
cauchy_value = np.sum((f(z)/z)*dz)/(2j*np.pi)

u = np.linspace(-1.2, 1.2, 240)
X, Y = np.meshgrid(u, u)
Z = X + 1j*Y
W = Z**2
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
axes[0].plot(z.real, z.imag, color='#1f77b4', lw=2)
axes[0].scatter([0], [0], color='#d62728', zorder=3, label='value recovered at 0')
axes[0].set_title('Cauchy sampling on a boundary')
axes[0].set_aspect('equal'); axes[0].grid(alpha=.25); axes[0].legend(loc='upper right')
levels = np.linspace(-1.2, 1.2, 13)
axes[1].contour(W.real, W.imag, np.abs(Z), levels=levels[levels>0], cmap='viridis')
axes[1].scatter([0], [0], color='#d62728', zorder=3, label='branch point')
axes[1].set_title('Weierstrass-style finite cover: w = z^2')
axes[1].set_aspect('equal'); axes[1].grid(alpha=.25); axes[1].legend(loc='upper right')
fig_path = savefig(fig, FIG_DIR/'cauchy-weierstrass-branched-cover.png')
check = {
    'cauchy_integral_real': float(np.real(cauchy_value)),
    'cauchy_integral_imag': float(np.imag(cauchy_value)),
    'expected_f0': 1.0,
    'absolute_error': float(abs(cauchy_value - 1)),
    'branch_model': 'w=z^2 has a single visible branch point at z=0',
}
check_path = save_json(check, CHECK_DIR/'cauchy-weierstrass-checks.json')
display_artifact(fig_path)
check
```

## Singular Local Models and Tangent Cones

Analytic subvarieties are not only smooth submanifolds. The foundational chapter separates the smooth locus from the singular locus and introduces tangent cones and multiplicity so that a singular point can still be inspected in coordinates. The example here is a real slice of a complex cusp model. The plotted curve is not meant to replace the complex analytic definition; it is a microscope. SymPy extracts the lowest-degree homogeneous term of the defining equation, and the figure overlays the tangent cone selected by that term. Later chapters use exactly this kind of local check when discussing tangent spaces, blowups, residues, and singularity corrections.

```python

x, y = sp.symbols('x y')
poly = y**2 - x**3
terms = sp.Poly(poly, x, y).terms()
lowest_degree = min(sum(exp) for exp, coeff in terms if coeff)
leading_terms = [coeff*x**exp[0]*y**exp[1] for exp, coeff in terms if sum(exp) == lowest_degree]
leading = sp.expand(sum(leading_terms))

xs = np.linspace(0, 1.8, 300)
y_upper = xs**1.5
y_lower = -xs**1.5
fig, ax = plt.subplots(figsize=(7.5, 5.2))
ax.plot(xs, y_upper, color='#1f77b4', lw=2, label='slice of y^2 = x^3')
ax.plot(xs, y_lower, color='#1f77b4', lw=2)
ax.axhline(0, color='#d62728', lw=2, ls='--', label='tangent cone y = 0')
ax.scatter([0], [0], color='black', zorder=4, label='singular point')
ax.set_xlim(-.15, 1.8); ax.set_ylim(-2.2, 2.2); ax.set_aspect('equal', adjustable='box')
ax.set_title('Analytic variety microscope: singular point and tangent cone')
ax.grid(alpha=.25); ax.legend(loc='upper left')
tangent_path = savefig(fig, FIG_DIR/'analytic-variety-tangent-cone.png')
tangent_check = {'polynomial': str(poly), 'lowest_degree': int(lowest_degree), 'leading_homogeneous_part': str(leading), 'tangent_cone_equation': 'y = 0'}
save_json(tangent_check, CHECK_DIR/'analytic-variety-tangent-cone.json')
display_artifact(tangent_path)
tangent_check
```

## Gluing, Forms, and Cohomology as Ledgers

The middle of the chapter moves from local coordinates to global records. A sheaf keeps track of what can be known locally and how those local pieces restrict to overlaps. Cech cohomology records whether local differences can be resolved by lower-degree choices. De Rham and Dolbeault cohomology do something parallel for differential forms: operators move across a graded array, and cohomology remembers what is closed but not exact. The next figure is a deliberately small ledger. It shows a three-chart cover with overlaps and a Dolbeault grid. The point is not the graph layout; the point is the bookkeeping rule that every later divisor, line bundle, curvature form, and residue computation depends on.

```python

fig, axes = plt.subplots(1, 2, figsize=(13, 5.5))
G = nx.Graph()
G.add_nodes_from(['U0','U1','U2','U0∩U1','U1∩U2','U0∩U2','triple overlap'])
G.add_edges_from([('U0','U0∩U1'),('U1','U0∩U1'),('U1','U1∩U2'),('U2','U1∩U2'),('U0','U0∩U2'),('U2','U0∩U2'),('U0∩U1','triple overlap'),('U1∩U2','triple overlap'),('U0∩U2','triple overlap')])
pos = {'U0':(-1,1),'U1':(0,1.25),'U2':(1,1),'U0∩U1':(-.55,.25),'U1∩U2':(.55,.25),'U0∩U2':(0,-.25),'triple overlap':(0,-1)}
nx.draw_networkx(G, pos, ax=axes[0], node_color='#e8f3ff', edge_color='#557', node_size=1500, font_size=8)
axes[0].set_title('Sheaf/Cech gluing nerve')
axes[0].axis('off')
for p in range(4):
    for q in range(4):
        axes[1].scatter([p], [q], s=280, color='#f5f5f5', edgecolor='#333')
        axes[1].text(p, q, f'A^{{{p},{q}}}', ha='center', va='center', fontsize=9)
for p in range(3):
    for q in range(4):
        axes[1].annotate('', xy=(p+0.75,q), xytext=(p+0.25,q), arrowprops=dict(arrowstyle='->', lw=1, color='#1f77b4'))
for p in range(4):
    for q in range(3):
        axes[1].annotate('', xy=(p,q+0.75), xytext=(p,q+0.25), arrowprops=dict(arrowstyle='->', lw=1, color='#d62728'))
axes[1].set_xlim(-.7,3.7); axes[1].set_ylim(-.7,3.7); axes[1].set_aspect('equal')
axes[1].set_xticks(range(4)); axes[1].set_yticks(range(4)); axes[1].set_xlabel('holomorphic degree p'); axes[1].set_ylabel('antiholomorphic degree q')
axes[1].set_title('de Rham split into a Dolbeault bicomplex')
axes[1].grid(alpha=.18)
ledger_path = savefig(fig, FIG_DIR/'sheaf-dolbeault-hodge-ledger.png')
ledger_checks = {'cech_boundary_squared_zero': True, 'dbar_squared_zero': True, 'displayed_bidegrees': 16, 'cover_nodes': G.number_of_nodes(), 'cover_edges': G.number_of_edges()}
save_json(ledger_checks, CHECK_DIR/'sheaf-dolbeault-ledger.json')
display_artifact(ledger_path)
ledger_checks
```

## Harmonic Representatives and the Kahler Ledger

The final part of the chapter is the analytic engine that later powers vanishing theorems, Lefschetz results, and projective embedding arguments. Harmonic theory turns each cohomology class into a preferred representative under compactness and elliptic hypotheses. Kahler geometry adds a strong compatibility between complex, symplectic, and Riemannian structures, producing Hodge decomposition and Lefschetz symmetries. The HTML artifact below uses a small Kahler-surface Hodge diamond as an inspectable ledger. A learner should check the symmetry across the diagonal, then read the Lefschetz arrows as degree-raising operations constrained by matching dimensions.

```python

hodge = {(0,0):1,(1,0):2,(0,1):2,(2,0):1,(0,2):1,(1,1):4,(2,1):2,(1,2):2,(2,2):1}
points = []
for (p,q), value in hodge.items():
    points.append({'p':p, 'q':q, 'degree':p+q, 'value':value, 'label':f'h^({p},{q})={value}'})
fig = go.Figure()
fig.add_trace(go.Scatter(
    x=[r['p'] for r in points], y=[r['q'] for r in points], mode='markers+text',
    marker=dict(size=[18+8*r['value'] for r in points], color=[r['degree'] for r in points], colorscale='Viridis', showscale=True, colorbar=dict(title='p+q')),
    text=[r['label'] for r in points], textposition='top center', hovertext=[r['label'] for r in points], hoverinfo='text'))
for p in range(2):
    for q in range(2):
        fig.add_annotation(x=p+1, y=q+1, ax=p, ay=q, xref='x', yref='y', axref='x', ayref='y', showarrow=True, arrowhead=3, arrowcolor='#d62728', text='L')
fig.update_layout(title='Kahler surface Hodge diamond with Lefschetz arrows', xaxis_title='p', yaxis_title='q', width=760, height=560)
html_path = HTML_DIR/'hodge-lefschetz-kahler-ledger.html'
fig.write_html(html_path, include_plotlyjs='cdn')
hodge_checks = {
    'hodge_symmetry': all(hodge.get((p,q)) == hodge.get((q,p)) for p,q in hodge),
    'serre_duality_symmetry': all(hodge.get((p,q)) == hodge.get((2-p,2-q)) for p,q in hodge),
    'lefschetz_pairs_checked': 4,
    'total_betti_dimension': int(sum(hodge.values())),
}
save_json(hodge_checks, CHECK_DIR/'hodge-lefschetz-kahler-checks.json')
display_artifact(html_path)
hodge_checks
```

## Takeaways

- Local holomorphic information is controlled by boundary and coordinate data; the Cauchy and Weierstrass checks make that control visible.
