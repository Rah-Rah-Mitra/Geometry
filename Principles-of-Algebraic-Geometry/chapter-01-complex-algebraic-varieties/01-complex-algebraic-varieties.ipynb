## Chapter Goal

This chapter starts turning the analytic foundations into algebraic geometry in projective space. The central idea is that local holomorphic and meromorphic data can be organized into global geometric objects: divisors, line bundles, Chern classes, projective embeddings, blowups, Grassmannians, and Schubert conditions. The notebook treats each of these as a data structure with a visible invariant. A divisor is a weighted record along hypersurfaces. A line bundle is transition data whose products must be compatible on overlaps. A projective variety has degree, tangent spaces, and hyperplane sections. A blowup replaces a point by the directions through it. A Grassmannian stores linear subspaces through Plucker coordinates and Schubert incidence conditions.

The chapter also contains several theorem engines: Kodaira vanishing, Lefschetz hyperplane results, Theorem B, the `(1,1)`-class statement, and the Kodaira embedding theorem. Full proofs require the analytic machinery built in Chapter 00. Here the proof-facing visuals record what the theorem takes as input and what output it certifies. That lets the reader separate the theorem's analytic hypothesis from the geometry it produces.

## Translation Guide

- Divisors become integer-weighted components with local equations and addition rules.
- Line bundles become transition functions on overlaps; the first Chern class is a curvature or cocycle ledger.
- Vanishing and Lefschetz theorems become dependency graphs: positivity plus compact Kahler input produces cohomological restrictions.
- Projective varieties become homogeneous equations with degree and tangent-space checks.
- The blowup is treated as a coordinate substitution and as a visible exceptional divisor of directions.
- Grassmannians become rank-two matrices modulo row operations, Plucker coordinates, cell dimensions, universal bundles, and Schubert incidence constraints.

## Divisors and Line Bundles as Compatible Local Data

A divisor keeps integer weights along codimension-one components. A line bundle remembers how local trivializations are glued. The two languages meet when a meromorphic section has zeros and poles and therefore produces divisor data. The figure below makes this concrete: three local charts carry transition functions on overlaps, and the product around the triple overlap is the cocycle test. The check is small, but it is the same bookkeeping that underlies Chern classes of line bundles and later Picard/Jacobian constructions.

```python

G=nx.DiGraph()
edges=[('U0','U1','g01=z'),('U1','U2','g12=1-z'),('U2','U0','g20=1/(z(1-z))')]
G.add_edges_from([(a,b,{'label':lab}) for a,b,lab in edges])
pos={'U0':(0,1),'U1':(-1,-.6),'U2':(1,-.6)}
fig, axes=plt.subplots(1,2,figsize=(12,5))
components=['H0: x=0','H1: y=0','H2: z=0']; weights=[2,-1,3]
axes[0].bar(components, weights, color=['#4c78a8','#f58518','#54a24b'])
axes[0].axhline(0,color='0.3',lw=1); axes[0].set_title('Divisor as weighted hypersurface ledger'); axes[0].set_ylabel('coefficient')
nx.draw_networkx_nodes(G,pos,ax=axes[1],node_size=1600,node_color='#e8f3ff',edgecolors='#333')
nx.draw_networkx_labels(G,pos,ax=axes[1])
nx.draw_networkx_edges(G,pos,ax=axes[1],arrows=True,arrowstyle='-|>',arrowsize=18,connectionstyle='arc3,rad=.12')
nx.draw_networkx_edge_labels(G,pos,{(a,b):lab for a,b,lab in edges},ax=axes[1],font_size=8)
axes[1].set_title('Line bundle transition cocycle'); axes[1].axis('off')
path=savefig(fig, FIG_DIR/'divisor-line-bundle-cocycle.png')
z=sp.symbols('z')
cocycle=sp.simplify(z*(1-z)*(1/(z*(1-z))))
divisor_checks={'divisor_degree':int(sum(weights)),'positive_part_degree':5,'negative_part_degree':1,'triple_overlap_cocycle':str(cocycle),'cocycle_is_one':bool(cocycle==1)}
save_json(divisor_checks, CHECK_DIR/'divisor-line-bundle-cocycle.json')
display_artifact(path)
divisor_checks
```

## Theorem Dependencies: Positivity into Projective Geometry

Vanishing, Lefschetz, Theorem B, the `(1,1)` statement, and Kodaira embedding are theorem engines rather than isolated definitions. Their role is to translate positivity and analytic hypotheses into cohomological disappearance, restriction maps, algebraicity of classes, and enough sections to map into projective space. The dependency graph is a proof visual: it does not prove the theorems, but it shows what must be available before each conclusion can be used safely.

```python

D=nx.DiGraph()
D.add_edges_from([
    ('positive line bundle','Kodaira vanishing'),('compact Kahler manifold','Kodaira vanishing'),('Kodaira vanishing','cohomology control'),
    ('hyperplane section','Lefschetz hyperplane theorem'),('cohomology control','Lefschetz hyperplane theorem'),
    ('closed (1,1)-class','Lefschetz (1,1) theorem'),('integral class','Lefschetz (1,1) theorem'),
    ('ample line bundle','Kodaira embedding'),('separating sections','Kodaira embedding'),('Kodaira embedding','projective model'),
])
pos=nx.spring_layout(D, seed=7, k=.75)
fig, ax=plt.subplots(figsize=(10,7))
nx.draw_networkx(D,pos,ax=ax,node_size=1700,node_color='#fff3cd',edge_color='#6c757d',font_size=8,arrows=True,arrowstyle='-|>',arrowsize=14)
ax.set_title('Vanishing, Lefschetz, and Kodaira dependency map'); ax.axis('off')
path=savefig(fig, FIG_DIR/'kodaira-lefschetz-dependency-map.png')
dep_checks={'nodes':D.number_of_nodes(),'edges':D.number_of_edges(),'projective_model_prerequisites':list(D.predecessors('Kodaira embedding')),'all_theorem_nodes_have_inputs':all(D.in_degree(n)>0 for n in ['Kodaira vanishing','Lefschetz hyperplane theorem','Lefschetz (1,1) theorem','Kodaira embedding'])}
save_json(dep_checks, CHECK_DIR/'kodaira-lefschetz-dependency-map.json')
display_artifact(path)
dep_checks
```

## Blowup Charts and the Exceptional Divisor

The blowup is one of the chapter's most important constructive moves. It does not merely erase a bad point; it replaces the point by the projective line of tangent directions through it. In one affine chart the map is `(u,v) -> (x=u, y=uv)`, so the exceptional divisor is `u=0` and the slope `v` remains. The plot shows several lines through the origin lifting to different points on the exceptional divisor. This same local maneuver reappears when surfaces are resolved and when intersection numbers are corrected after blowup.

```python

slopes=np.array([-2,-1,-.4,.4,1,2])
t=np.linspace(-1.0,1.0,80)
fig, axes=plt.subplots(1,2,figsize=(12,5))
for m in slopes:
    axes[0].plot(t, m*t, lw=1.6, label=f'slope {m:g}')
axes[0].scatter([0],[0],color='black',zorder=5); axes[0].set_title('Lines meeting at the point to be blown up')
axes[0].set_aspect('equal'); axes[0].grid(alpha=.25)
axes[1].axvline(0,color='#d62728',lw=3,label='exceptional divisor u=0')
for m in slopes:
    axes[1].scatter([0],[m],s=60); axes[1].text(.04,m,f'v={m:g}',va='center')
axes[1].set_xlim(-.4,1.2); axes[1].set_ylim(-2.5,2.5); axes[1].set_xlabel('u'); axes[1].set_ylabel('slope coordinate v')
axes[1].set_title('Blowup chart: each direction survives as a point'); axes[1].grid(alpha=.25); axes[1].legend()
path=savefig(fig, FIG_DIR/'blowup-exceptional-chart.png')
u,v=sp.symbols('u v')
x,y=sp.symbols('x y')
chart_equation=sp.simplify(y-u*v)
blowup_checks={'chart_map':'x=u, y=u*v','exceptional_equation':'u=0','sampled_directions':slopes.tolist(),'chart_equation_after_substitution':str(chart_equation.subs({y:u*v}))}
save_json(blowup_checks, CHECK_DIR/'blowup-exceptional-chart.json')
display_artifact(path)
blowup_checks
```

## Grassmannians, Schubert Cells, and Plucker Coordinates

The Grassmannian section gives a concrete parameter space for linear subspaces. A two-plane in four-space can be represented by a `2 x 4` matrix up to row operations. Its `2 x 2` minors are Plucker coordinates, and those coordinates are not arbitrary: for `G(2,4)` they satisfy the Klein quadric relation. Schubert cells add incidence conditions, turning the Grassmannian into a stratified geometric object. The static figure records the cell dimensions and a sample two-plane; the HTML explorer stores the Plucker coordinates and exact relation so that the learner can inspect the algebraic embedding rather than only name it.

```python

A=sp.Matrix([[1,0,2,1],[0,1,-1,3]])
cols=[(0,1),(0,2),(0,3),(1,2),(1,3),(2,3)]
plucker={f'p{i+1}{j+1}':sp.Matrix(A[:,[i,j]]).det() for i,j in cols}
klein=sp.simplify(plucker['p12']*plucker['p34'] - plucker['p13']*plucker['p24'] + plucker['p14']*plucker['p23'])
cell_dims=np.array([[0,1,2],[1,2,3],[2,3,4]])
fig, axes=plt.subplots(1,2,figsize=(12,5))
im=axes[0].imshow(cell_dims,cmap='Blues')
for i in range(cell_dims.shape[0]):
    for j in range(cell_dims.shape[1]): axes[0].text(j,i,str(cell_dims[i,j]),ha='center',va='center')
axes[0].set_title('Schubert cell dimension ledger'); axes[0].set_xlabel('incidence step'); axes[0].set_ylabel('rank condition')
fig.colorbar(im,ax=axes[0],fraction=.046)
axes[1].bar(list(plucker.keys()), [float(v) for v in plucker.values()], color='#4c78a8')
axes[1].axhline(0,color='0.3',lw=1); axes[1].set_title('Sample Plucker coordinates for a 2-plane')
path=savefig(fig, FIG_DIR/'grassmannian-schubert-plucker-ledger.png')
fig_html=go.Figure(data=[go.Bar(x=list(plucker.keys()), y=[float(v) for v in plucker.values()], text=[str(v) for v in plucker.values()], textposition='auto')])
fig_html.update_layout(title='Plucker coordinate explorer: Klein relation stays zero', yaxis_title='minor value')
html_path=HTML_DIR/'plucker-coordinate-explorer.html'
fig_html.write_html(html_path, include_plotlyjs='cdn')
plucker_checks={'matrix':[[int(A[i,j]) for j in range(A.shape[1])] for i in range(A.shape[0])],'plucker':{k:int(v) for k,v in plucker.items()},'klein_relation':str(klein),'klein_relation_zero':bool(klein==0),'schubert_dimension_max':int(cell_dims.max())}
save_json(plucker_checks, CHECK_DIR/'grassmannian-schubert-plucker-ledger.json')
display_artifact(path); display_artifact(html_path)
plucker_checks
```

## Takeaways

- Divisors and line bundles are two ways to encode local-to-global hypersurface data; compatibility on overlaps is the recurring check.
- Vanishing and Lefschetz results are best remembered by their inputs and outputs: positivity and Kahler hypotheses produce cohomological and projective conclusions.
- Blowing up replaces a point by its directions, making tangent data visible and preparing the surface theory later in the book.
- Grassmannians are concrete algebraic varieties: Schubert cells stratify them, universal bundles live over them, and Plucker coordinates embed them into projective space subject to exact relations.
