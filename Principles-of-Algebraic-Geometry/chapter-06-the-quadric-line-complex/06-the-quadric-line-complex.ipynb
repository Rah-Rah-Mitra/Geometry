## Quadrics and Linear Systems: Rank as the First Invariant

A quadric in projective space is controlled by a symmetric matrix up to scale. Rank records the first geometric degeneration: smooth quadrics, cones, pairs of planes, and lower-rank cases sit in different strata. A linear system of quadrics is then a family of such matrices. The determinant of a pencil marks where the rank drops. The figure below keeps those two ideas together: diagonal rank samples on one side, determinant values for a pencil on the other.

```python

t=sp.symbols('t')
Q0=sp.diag(1,1,1,1)
Q1=sp.diag(1,-1,2,-2)
pencil=Q0+t*Q1
det_poly=sp.factor(pencil.det())
samples=np.linspace(-1.4,1.4,29)
ranks=[]; det_vals=[]
for s in samples:
    M=np.diag([1+s,1-s,1+2*s,1-2*s])
    ranks.append(int(np.linalg.matrix_rank(M,tol=1e-9)))
    det_vals.append(float(np.linalg.det(M)))
rank_grid=np.array(ranks).reshape(1,-1)
fig,axes=plt.subplots(2,1,figsize=(11,6),gridspec_kw={'height_ratios':[1,2]})
im=axes[0].imshow(rank_grid,aspect='auto',cmap='viridis',vmin=1,vmax=4)
axes[0].set_yticks([]); axes[0].set_xticks(range(0,len(samples),4),[f'{samples[i]:.1f}' for i in range(0,len(samples),4)]); axes[0].set_title('Rank strata in a diagonal pencil')
fig.colorbar(im,ax=axes[0],orientation='horizontal',fraction=.2,pad=.25,label='rank')
axes[1].plot(samples,det_vals,marker='o',ms=3); axes[1].axhline(0,color='#d62728',ls='--'); axes[1].set_xlabel('pencil parameter t'); axes[1].set_ylabel('det(Q0+tQ1)'); axes[1].set_title('Discriminant values for the linear system'); axes[1].grid(alpha=.25)
path=savefig(fig,FIG_DIR/'quadric-rank-pencil-discriminant.png')
rank_checks={'determinant_polynomial':str(det_poly),'sampled_ranks':ranks,'rank_drops':int(sum(r<4 for r in ranks)),'determinant_roots':[str(r) for r in sp.solve(sp.Eq(det_poly,0),t)]}
save_json(rank_checks,CHECK_DIR/'quadric-rank-pencil-discriminant.json')
display_artifact(path)
rank_checks
```

## Lines in `P^3`: Plucker Coordinates and the Klein Relation

The Grassmannian `G(2,4)` parametrizes projective lines in `P^3`. Given two homogeneous points, the six `2 x 2` minors are Plucker coordinates. They must satisfy the Klein quadric relation, so line geometry becomes algebraic geometry in projective Plucker space. A line complex adds another equation. The static artifact records sampled line coordinates; the HTML artifact lets the learner inspect the line-complex values alongside the exact Klein residuals.

```python

def plucker(a,b):
    vals=[]
    for i,j in [(0,1),(0,2),(0,3),(1,2),(1,3),(2,3)]: vals.append(a[i]*b[j]-a[j]*b[i])
    return np.array(vals,dtype=float)
lines=[(np.array([1,0,0,0.0]),np.array([0,1,1,0.0])),(np.array([1,1,0,0.0]),np.array([0,0,1,1.0])),(np.array([1,0,1,0.0]),np.array([0,1,0,1.0])),(np.array([1,2,1,0.0]),np.array([0,1,2,1.0]))]
labels=['p12','p13','p14','p23','p24','p34']
P=np.array([plucker(a,b) for a,b in lines])
klein=P[:,0]*P[:,5]-P[:,1]*P[:,4]+P[:,2]*P[:,3]
complex_values=P@np.array([1,-1,0.5,0,1,0.25])
fig,axes=plt.subplots(1,2,figsize=(13,5))
im=axes[0].imshow(P,cmap='coolwarm'); axes[0].set_xticks(range(6),labels); axes[0].set_yticks(range(len(lines)),[f'L{i}' for i in range(len(lines))]); axes[0].set_title('Sample Plucker coordinate ledger')
for i in range(P.shape[0]):
    for j in range(P.shape[1]): axes[0].text(j,i,f'{P[i,j]:.0f}',ha='center',va='center')
axes[1].bar([f'L{i}' for i in range(len(lines))],klein,color='#54a24b',label='Klein residual'); axes[1].plot([f'L{i}' for i in range(len(lines))],complex_values,marker='o',color='#f58518',label='line-complex value'); axes[1].axhline(0,color='0.3',lw=1); axes[1].legend(); axes[1].set_title('Klein relation and line-complex equation')
path=savefig(fig,FIG_DIR/'klein-plucker-line-complex-ledger.png')
fig_html=go.Figure()
for idx,row in enumerate(P): fig_html.add_trace(go.Bar(x=labels,y=row,name=f'L{idx}: complex={complex_values[idx]:.2f}'))
fig_html.update_layout(title='Plucker line-complex explorer: all sampled lines lie on the Klein quadric',barmode='group',yaxis_title='coordinate value')
html_path=HTML_DIR/'plucker-line-complex-explorer.html'; fig_html.write_html(html_path,include_plotlyjs='cdn')
plucker_checks={'plucker_labels':labels,'klein_residuals':klein.tolist(),'klein_all_zero':bool(np.allclose(klein,0)),'line_complex_values':complex_values.tolist()}
save_json(plucker_checks,CHECK_DIR/'klein-plucker-line-complex-ledger.json')
display_artifact(path); display_artifact(html_path)
plucker_checks
```

## Kummer Surface and Singular-Line Configurations as Incidence Data

The chapter's Kummer and configuration discussions are classical and visual, but copying a textbook figure would not help execution. The graph below builds a small incidence scaffold: sixteen node labels are grouped into four blocks, with additional links representing singular-line or trope-style relationships. The invariant is combinatorial: each block has a fixed internal incidence count and the cross-links are recorded explicitly. This gives a learner a modifiable representation for configuration arguments.

```python

G=nx.Graph()
blocks=[]
for b in range(4):
    block=[f'N{b}{i}' for i in range(4)]; blocks.append(block); G.add_nodes_from(block,block=b)
    for i in range(4): G.add_edge(block[i],block[(i+1)%4],kind='block')
for i in range(4):
    G.add_edge(blocks[0][i],blocks[1][i],kind='cross'); G.add_edge(blocks[2][i],blocks[3][(i+1)%4],kind='cross')
pos=nx.circular_layout(G)
colors=[G.nodes[n]['block'] for n in G.nodes]
fig,ax=plt.subplots(figsize=(8,8)); nx.draw_networkx(G,pos,ax=ax,node_color=colors,cmap='Set3',edge_color=['#555' if G.edges[e]['kind']=='block' else '#d62728' for e in G.edges],node_size=900,font_size=8)
ax.set_title('Kummer-style singular-line configuration scaffold'); ax.axis('off')
path=savefig(fig,FIG_DIR/'kummer-singular-line-configuration-graph.png')
block_edge_counts=[]
for block in blocks:
    sub=G.subgraph(block); block_edge_counts.append(sub.number_of_edges())
config_checks={'node_count':G.number_of_nodes(),'edge_count':G.number_of_edges(),'block_edge_counts':block_edge_counts,'each_block_cycle_has_four_edges':all(c==4 for c in block_edge_counts)}
save_json(config_checks,CHECK_DIR/'kummer-singular-line-configuration-graph.json')
display_artifact(path)
config_checks
```

## Group Law and Rationality Parameter Scaffold

The line variety and reprise sections use a group-law flavor and rationality arguments. The notebook uses a simple rational parameter map as a scaffold: points are sampled from a parameter, an operation is checked by residuals on the model equation, and exceptional parameter values are recorded rather than hidden. This is not the theorem; it is an executable model for the habit of tracking parameters, exceptional loci, and algebraic residuals.

```python

ts=np.linspace(-2,2,200)
ts=ts[np.abs(ts-1)>0.08]
X=(ts**2-1)/(ts**2+1); Y=2*ts/(ts**2+1)
residual=X**2+Y**2-1
sample_ops=[(-.5,.75),(0.25,1.2)]
op_residuals=[]
for a,b in sample_ops:
    # tangent-addition style operation on circle parameters
    c=(a+b)/(1-a*b) if abs(1-a*b)>1e-9 else np.inf
    if np.isfinite(c):
        x=(c*c-1)/(c*c+1); y=2*c/(c*c+1); op_residuals.append(float(x*x+y*y-1))
fig,axes=plt.subplots(1,2,figsize=(12,5))
axes[0].plot(X,Y,color='#4c78a8'); axes[0].scatter(X[::25],Y[::25],color='#f58518'); axes[0].set_aspect('equal'); axes[0].grid(alpha=.25); axes[0].set_title('Rational parameter map to a model curve')
axes[1].plot(ts,residual); axes[1].axhline(0,color='#d62728',ls='--'); axes[1].set_xlabel('parameter t'); axes[1].set_ylabel('equation residual'); axes[1].set_title('Residual stays zero away from exceptional values'); axes[1].grid(alpha=.25)
path=savefig(fig,FIG_DIR/'group-law-rationality-parameter-map.png')
param_checks={'model_equation':'x^2 + y^2 = 1','max_parameter_residual':float(np.max(np.abs(residual))),'operation_residuals':op_residuals,'operation_residuals_small':bool(all(abs(r)<1e-12 for r in op_residuals)),'exceptional_parameters':['t = infinity for one chart','a*b = 1 for the sampled operation formula']}
save_json(param_checks,CHECK_DIR/'group-law-rationality-parameter-map.json')
display_artifact(path)
param_checks
```
