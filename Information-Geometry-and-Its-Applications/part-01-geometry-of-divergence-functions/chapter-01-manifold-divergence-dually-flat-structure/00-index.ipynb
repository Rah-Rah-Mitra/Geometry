## Focus

Manifolds, asymmetric divergence, Bregman geometry, Legendre duality, affine coordinates, and generalized Pythagorean projection.

## Concepts to Rebuild in Original Prose

- a coordinate chart is a computational handle, not the manifold itself
- a divergence compares two points without requiring symmetry
- a convex potential creates dual coordinates through the Legendre transform
- orthogonal e- and m-projections turn the Pythagorean theorem into an algorithm

## Computational Checks

- Bregman divergence is non-negative and zero at identical points
- Legendre gradients invert on the plotted convex potential
- m-geodesic probabilities stay normalized
- Pythagorean residual is small for the constructed orthogonal projection
