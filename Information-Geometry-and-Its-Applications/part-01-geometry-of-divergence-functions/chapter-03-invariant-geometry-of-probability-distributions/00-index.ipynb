## Concepts to Rebuild in Original Prose

- coarse graining cannot increase distinguishability for invariant divergences
- f-divergences share a common convex-generator form
- Fisher information is the local metric selected by invariance
- positive measures widen the geometry beyond normalized distributions

## Computational Checks

- coarse-grained KL does not exceed fine KL
- f-divergence is zero for identical distributions
- categorical Fisher metric is symmetric positive semidefinite
- random stochastic matrices preserve total probability
