## Focus

EM as alternating divergence minimization, hidden-variable models, Gaussian mixtures, information loss under data reduction, and misspecification.

## Concepts to Rebuild in Original Prose

- hidden variables enlarge the complete-data manifold
- EM alternates between compatible observed-data and model constraints
- data reduction loses information when the statistic is insufficient
- misspecification replaces truth with the closest model point

## Computational Checks

- responsibilities sum to one for each observation
- EM log-likelihood is non-decreasing in the demonstration
- coarse-grained divergence is no larger than original divergence
- misspecified projection minimizes the sampled KL curve
