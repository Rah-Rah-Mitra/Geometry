Source orientation: printed pages 109-130; PDF pages 126-147.

## Focus

Manifolds, tangent spaces, metrics, affine connections, tensors, covariant derivatives, geodesics, curvature, Levi-Civita connection, and submanifolds.

## Concepts to Rebuild in Original Prose

- tangent vectors are directional derivatives expressed in a chart
- a metric turns local displacement into measurable length
- an affine connection tells vectors how to move between tangent spaces
- curvature records path dependence of transport

## Visual Storyboard

- coordinate chart and tangent basis on a curved surface
- metric ellipse field for a non-Cartesian coordinate system
- geodesic shooting compared with straight chart lines
- round-trip parallel transport revealing curvature

## Computational Checks

- metric matrices are symmetric positive definite on sampled points
- geodesic integration preserves expected first-order speed
- Levi-Civita symbols are torsion-free in the example
- parallel transport around a flat loop returns the vector
