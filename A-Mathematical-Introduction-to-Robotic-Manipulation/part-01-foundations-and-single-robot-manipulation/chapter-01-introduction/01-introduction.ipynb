## Worked Interpretation

The dependency graph is a worked example: rigid motion supports product-of-exponentials kinematics, Jacobians support dynamics and grasping, and nonholonomic distributions support motion planning. Changing one node in that graph changes what later chapters can assume. This is the chapter's small worked example, not a full simulator. It is intentionally narrow enough that the relevant convention can be seen directly in code and broad enough to catch the common failure mode.

The visual sequence and the executable check use compatible parameters whenever possible. The point is to avoid a split course where the images tell one story and the numbers tell another. When extending this notebook, reuse that pattern: pick one invariant, draw the geometry that exposes it, then save a check payload that would fail if the convention were reversed or the rank condition were lost.

**Pitfall to watch:** A common reading mistake is to treat the history as detachable from the mathematics. In this course, the history explains why open chains, contact, sensing, and velocity constraints all need one coherent language rather than separate ad hoc coordinate tricks. This pitfall is why the notebook avoids silent coordinate conventions. Names, frames, dimensions, and signs are part of the teaching surface, not implementation trivia.


```python
import networkx as nx

dependencies = nx.DiGraph()
dependencies.add_edges_from([
    ("rigid motion", "manipulator kinematics"),
    ("manipulator kinematics", "dynamics"),
    ("manipulator kinematics", "grasping"),
    ("grasping", "hand control"),
    ("rigid motion", "nonholonomic systems"),
    ("nonholonomic systems", "motion planning"),
])
check_payload = {
    "dependency_graph_acyclic": nx.is_directed_acyclic_graph(dependencies),
    "course_nodes": dependencies.number_of_nodes(),
    "course_edges": dependencies.number_of_edges(),
    "sample_serial_chain_dof": 6,
}
check_payload
```
