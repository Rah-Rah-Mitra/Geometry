## Translation Guide

- **Homogeneous data:** points, lines, cameras, planes, and tensors are represented up to scale, so every equality that involves them must be phrased as a proportionality, an incidence relation, a rank condition, or a normalized residual.
- **Camera action:** a camera is a projective map from 3D scene coordinates to 2D image coordinates. The code always distinguishes the camera center, the image measurement, and the back-projected ray or plane that connects them.
- **Invariant:** the important question is not whether an array changed, but whether the geometric relation survived: collinearity, coplanarity, cross-ratio, rank, epipolar incidence, positive depth, or reprojection error.
- **Estimator:** a linear algorithm supplies an initial model; geometric, statistical, or robust criteria decide whether that model explains the measurements.
- **Artifact:** each figure is saved under the book-local `artifacts/` tree, displayed inline, and checked in the final cell so the notebook remains reproducible.

## Worked Example Pattern

The worked examples use a shared synthetic lab. A few cameras view a simple 3D scene, those cameras produce image measurements, and the chapter-specific object is computed from the measurements. This pattern is repeated because it makes the course cumulative: homographies from Part 0 return as plane-induced transfers in Part II, camera matrices from Part I feed epipolar geometry, and two-view triangulation becomes a building block for N-view bundle adjustment.

For this chapter, the important work is to connect **Covariant and contravariant indices, basis changes, contractions, the epsilon tensor, and trifocal tensor notation.** to observable behavior. The first figure is a concept map that states what to watch. The second figure is a geometry scene. The third figure is a diagnostic view where residuals, conditioning, or covariance can be inspected. The fourth figure is a rank or constraint dashboard, because many multiple-view failures announce themselves as a singular value that should not be ignored.

The code is intentionally compact. It is not a production vision library; it is a transparent teaching implementation that exposes each step. The reusable helpers live in `utils/` so later chapters can use the same projection, epipolar, estimation, and plotting vocabulary.

```python
import json
import math

import matplotlib.pyplot as plt
import numpy as np

from utils.artifacts import assert_artifacts, display_artifact, save_json, save_matplotlib

ENTRY_TITLE = 'Tensor Notation'
MODE = 'tensor-notation'
TOPIC = 'appendix-01'
CONCEPTS = ['contravariant coordinates', 'covariant components', 'basis change', 'contraction', 'epsilon tensor', 'trifocal slices']
VISUALS = ['basis-change index diagram', 'epsilon determinant contraction', 'tensor slice contraction map', 'trifocal notation flow']
CHECKS = ['basis-transformed coordinates represent the same point', 'contracted scalar values remain invariant', 'epsilon contractions match determinants', 'tensor slice contraction agrees with einsum']
SEED = 701
rng = np.random.default_rng(SEED)
artifact_paths = []

A = np.array([[1.35, 0.42, -0.10], [0.15, 0.95, 0.28], [0.0, -0.22, 1.10]])
v = np.array([0.8, -1.1, 0.6])
alpha = np.array([1.2, -0.4, 0.9])
v_new = np.linalg.solve(A, v)
alpha_new = A.T @ alpha
scalar_old = float(alpha @ v)
scalar_new = float(alpha_new @ v_new)

def levi_civita_3():
    eps = np.zeros((3, 3, 3), dtype=int)
    for i in range(3):
        for j in range(3):
            for k in range(3):
                perm = [i, j, k]
                if len(set(perm)) < 3:
                    continue
                inversions = sum(perm[a] > perm[b] for a in range(3) for b in range(a + 1, 3))
                eps[i, j, k] = -1 if inversions % 2 else 1
    return eps

epsilon = levi_civita_3()
a = np.array([1.4, -0.2, 0.7])
b = np.array([-0.5, 1.1, 0.3])
cross_eps = np.einsum('ijk,j,k->i', epsilon, a, b)
cross_np = np.cross(a, b)
T = rng.normal(size=(3, 3, 3))
line = np.array([0.5, -0.7, 1.0])
point = np.array([1.0, 0.35, -0.2])
contracted = np.einsum('ijk,j,k->i', T, line, point)
manual = np.array([sum(T[i, j, k] * line[j] * point[k] for j in range(3) for k in range(3)) for i in range(3)])
```

```python
fig, axes = plt.subplots(1, 3, figsize=(10.4, 3.7))
for ax, matrix, title in [(axes[0], A, 'basis matrix A'), (axes[1], np.column_stack([v, v_new]), 'vector coordinates'), (axes[2], np.column_stack([alpha, alpha_new]), 'covector components')]:
    im = ax.imshow(matrix, cmap='coolwarm')
    ax.set_title(title, loc='left', fontsize=10, fontweight='bold')
    ax.set_xticks(range(matrix.shape[1])); ax.set_yticks(range(matrix.shape[0]))
    for (i, j), value in np.ndenumerate(matrix):
        ax.text(j, i, f'{value:.2f}', ha='center', va='center', fontsize=8)
    fig.colorbar(im, ax=ax, fraction=0.046, pad=0.04)
fig.suptitle('Contravariant coordinates and covariant components move oppositely under basis change', x=0.02, ha='left', fontsize=12, fontweight='bold')
fig.tight_layout()
basis_path = save_matplotlib(fig, TOPIC, 'figures', 'basis-change-index-diagram.png')
plt.close(fig)
artifact_paths.append(basis_path)
display_artifact(basis_path, width=900)
```

```python
fig = plt.figure(figsize=(8.4, 5.2))
ax = fig.add_subplot(111, projection='3d')
origin = np.zeros(3)
for vec, color, label in [(a, '#2563eb', 'a'), (b, '#0f766e', 'b'), (cross_eps, '#c2410c', 'epsilon contraction')]:
    ax.quiver(*origin, *vec, color=color, linewidth=2.2, arrow_length_ratio=0.12, label=label)
par = np.array([origin, a, a + b, b, origin])
ax.plot(par[:, 0], par[:, 1], par[:, 2], color='#94a3b8', lw=1.5, label='oriented parallelogram')
ax.set_title('Levi-Civita tensor encodes the signed cross product', loc='left', fontsize=12, fontweight='bold')
ax.set_xlabel('x'); ax.set_ylabel('y'); ax.set_zlabel('z')
ax.legend(loc='upper left', fontsize=8)
ax.view_init(elev=23, azim=-48)
epsilon_path = save_matplotlib(fig, TOPIC, 'figures', 'epsilon-determinant-contraction.png')
plt.close(fig)
artifact_paths.append(epsilon_path)
display_artifact(epsilon_path, width=820)
```

```python
fig, axes = plt.subplots(1, 4, figsize=(11.2, 3.2))
for i in range(3):
    im = axes[i].imshow(T[i], cmap='viridis')
    axes[i].set_title(f'T[{i}, :, :]', loc='left', fontsize=10, fontweight='bold')
    axes[i].set_xticks(range(3)); axes[i].set_yticks(range(3))
    fig.colorbar(im, ax=axes[i], fraction=0.046, pad=0.04)
axes[3].bar(np.arange(3), contracted, color=['#2563eb', '#0f766e', '#c2410c'])
axes[3].axhline(0, color='#334155', lw=0.8)
axes[3].set_title('contracted output', loc='left', fontsize=10, fontweight='bold')
axes[3].set_xticks(range(3), ['i=0', 'i=1', 'i=2'])
fig.suptitle('A third-order tensor is a stack of matrices; contraction weights the slices', x=0.02, ha='left', fontsize=12, fontweight='bold')
fig.tight_layout()
slice_path = save_matplotlib(fig, TOPIC, 'figures', 'tensor-slice-contraction-map.png')
plt.close(fig)
artifact_paths.append(slice_path)
display_artifact(slice_path, width=900)
```

## Pitfalls And Failure Modes

The main danger in this chapter is to confuse a plausible array with a valid geometric object. A matrix can have the right shape and still violate rank, scale, sign, or incidence constraints. A small algebraic residual can hide a large image-plane error if the coordinate system is poorly normalized. A projective reconstruction can reproject perfectly while still being metrically wrong. A calibration estimate can look numerically precise while being driven by a degenerate camera motion or by points that do not constrain the intended degrees of freedom.

The antidote is to make each assumption executable. When a relation is homogeneous, normalize before comparing. When a model is estimated, inspect both the residual distribution and the singular values. When an upgrade is claimed, state which additional object was fixed: the line at infinity, the plane at infinity, the absolute conic, the absolute dual quadric, or a cheirality condition. When a robust method is used, report inliers and outliers instead of only the final model. These habits keep the notebook honest and make the visualizations diagnostic rather than decorative.
