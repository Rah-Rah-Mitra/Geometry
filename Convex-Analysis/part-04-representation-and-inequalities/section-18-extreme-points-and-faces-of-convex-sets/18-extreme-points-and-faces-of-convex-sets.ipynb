## Translation Guide

Part IV turns finite-dimensional structure into representation theorems and certificates for inequalities. The notebooks keep the dimension count visible through simplices, faces, active constraints, and alternatives.

For this section, the translation into computational language is centered on extreme point, face, exposed face, supporting hyperplane. The notebook treats those ideas as objects with coordinates, inequalities, sampled witnesses, and explicit residuals. That is deliberately narrower than the full theorem set of the printed section, but it is broad enough to make the main geometry visible and to support later refinement by a section worker.

- `extreme point`: a point that cannot be split as a nontrivial segment average inside the set.
- `face`: a convex subset that contains every segment interior point it captures.
- `exposed face`: the points where a support functional is maximized.
- `supporting hyperplane`: a hyperplane that touches the set without cutting through it.
