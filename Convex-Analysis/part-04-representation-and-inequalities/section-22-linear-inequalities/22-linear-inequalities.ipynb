## Translation Guide

Part IV turns finite-dimensional structure into representation theorems and certificates for inequalities. The notebooks keep the dimension count visible through simplices, faces, active constraints, and alternatives.

For this section, the translation into computational language is centered on Farkas certificate, linear system, cone membership, alternative theorem. The notebook treats those ideas as objects with coordinates, inequalities, sampled witnesses, and explicit residuals. That is deliberately narrower than the full theorem set of the printed section, but it is broad enough to make the main geometry visible and to support later refinement by a section worker.

- `Farkas certificate`: a separating vector proving a linear alternative.
- `linear system`: equations and inequalities expressed by matrix rows.
- `cone membership`: whether a vector is a nonnegative combination of generators.
- `alternative theorem`: a statement that exactly one of two certificate systems is possible.
