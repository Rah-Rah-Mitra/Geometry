## Translation Guide

Part IV turns finite-dimensional structure into representation theorems and certificates for inequalities. The notebooks keep the dimension count visible through simplices, faces, active constraints, and alternatives.

For this section, the translation into computational language is centered on Helly theorem, intersection certificate, systems of inequalities, dimension. The notebook treats those ideas as objects with coordinates, inequalities, sampled witnesses, and explicit residuals. That is deliberately narrower than the full theorem set of the printed section, but it is broad enough to make the main geometry visible and to support later refinement by a section worker.

- `Helly theorem`: the principle that low-dimensional subfamilies can certify intersection.
- `intersection certificate`: a point or finite subfamily that proves feasibility or infeasibility.
- `systems of inequalities`: families of constraints read as intersections of convex sets.
- `dimension`: the affine dimension of the set being inspected.
