## Translation Guide

Part II asks what convexity does to boundary behavior. The visuals emphasize relative topology, closure, recession, and continuity as geometric constraints rather than as detached epsilon statements.

For this section, the translation into computational language is centered on linear image, projection, closedness, recession obstruction. The notebook treats those ideas as objects with coordinates, inequalities, sampled witnesses, and explicit residuals. That is deliberately narrower than the full theorem set of the printed section, but it is broad enough to make the main geometry visible and to support later refinement by a section worker.

- `linear image`: a projection or transformation whose closedness can fail without hypotheses.
- `projection`: a linear map that forgets coordinates and may lose limiting boundary points.
- `closedness`: topological completeness of a set after the operation being studied.
- `recession obstruction`: a direction at infinity that explains a closedness failure.
