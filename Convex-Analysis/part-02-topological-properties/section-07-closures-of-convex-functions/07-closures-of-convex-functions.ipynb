## Translation Guide

Part II asks what convexity does to boundary behavior. The visuals emphasize relative topology, closure, recession, and continuity as geometric constraints rather than as detached epsilon statements.

For this section, the translation into computational language is centered on lower semicontinuity, closed epigraph, closure operation, boundary value. The notebook treats those ideas as objects with coordinates, inequalities, sampled witnesses, and explicit residuals. That is deliberately narrower than the full theorem set of the printed section, but it is broad enough to make the main geometry visible and to support later refinement by a section worker.

- `lower semicontinuity`: the no-downward-jump condition that corresponds to closed epigraphs.
- `closed epigraph`: an epigraph containing its limiting boundary points.
- `closure operation`: the operation that repairs a convex function at missing boundary values.
- `boundary value`: a value forced by limiting behavior from inside the domain.
