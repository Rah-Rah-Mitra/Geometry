## Translation Guide

Part V studies how nonsmooth convex functions still carry derivative-like information. The section visuals use supporting lines, monotone graphs, and gradient maps to make subdifferentiation inspectable.

For this section, the translation into computational language is centered on gradient, differentiability, smooth convex function, nonsmooth point. The notebook treats those ideas as objects with coordinates, inequalities, sampled witnesses, and explicit residuals. That is deliberately narrower than the full theorem set of the printed section, but it is broad enough to make the main geometry visible and to support later refinement by a section worker.

- `gradient`: the ordinary derivative vector when it exists.
- `differentiability`: the collapse of the subdifferential to one vector.
- `smooth convex function`: a convex function whose gradient varies continuously in the region inspected.
- `nonsmooth point`: a point where the supporting slopes do not collapse to one vector.
