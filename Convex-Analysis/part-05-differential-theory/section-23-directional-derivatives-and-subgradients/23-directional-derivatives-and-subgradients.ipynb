## Translation Guide

Part V studies how nonsmooth convex functions still carry derivative-like information. The section visuals use supporting lines, monotone graphs, and gradient maps to make subdifferentiation inspectable.

For this section, the translation into computational language is centered on directional derivative, subgradient, subdifferential, supporting line. The notebook treats those ideas as objects with coordinates, inequalities, sampled witnesses, and explicit residuals. That is deliberately narrower than the full theorem set of the printed section, but it is broad enough to make the main geometry visible and to support later refinement by a section worker.

- `directional derivative`: the one-sided rate of change in a chosen direction.
- `subgradient`: a slope vector defining a global supporting affine function.
- `subdifferential`: the set of all subgradients at a point.
- `supporting line`: a one-dimensional supporting hyperplane for a function graph or epigraph.
