## Translation Guide

Part VI moves from geometry to optimization. The notebooks show minimizers, multipliers, perturbations, dual programs, and Fenchel equality as visible balance conditions.

For this section, the translation into computational language is centered on minimizer, level set, recession, zero subgradient. The notebook treats those ideas as objects with coordinates, inequalities, sampled witnesses, and explicit residuals. That is deliberately narrower than the full theorem set of the printed section, but it is broad enough to make the main geometry visible and to support later refinement by a section worker.

- `minimizer`: a point where the convex objective reaches its lowest value.
- `level set`: a slice where a function is bounded by a chosen value.
- `recession`: escape behavior that controls whether minima exist.
- `zero subgradient`: the first-order optimality condition for unconstrained convex minimization.
