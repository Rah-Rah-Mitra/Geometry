## Translation Guide

Part I builds the dictionary between finite-dimensional geometry and convex functions. The recurring move is to turn algebraic closure rules into pictures that can be checked with points, half-spaces, hulls, and epigraphs.

For this section, the translation into computational language is centered on convex combination, half-space, convex cone, polytope. The notebook treats those ideas as objects with coordinates, inequalities, sampled witnesses, and explicit residuals. That is deliberately narrower than the full theorem set of the printed section, but it is broad enough to make the main geometry visible and to support later refinement by a section worker.

- `convex combination`: a weighted average with nonnegative weights summing to one.
- `half-space`: one side of a hyperplane, expressed by a linear inequality.
- `convex cone`: a set closed under positive scaling and addition.
- `polytope`: a convex hull of finitely many points.
