## Translation Guide

- Global surface geometry is controlled by how geodesics extend and minimize.
- Jacobi fields measure the stretching of geodesic polar circles.
- Covering surfaces separate local geometry from global topology.
- Curvature inequalities force strong global conclusions.

A computational translation is not a replacement for the definition. It is a second view of the same definition. In this notebook, formulas are kept close to the object they measure. A vector field is tested pointwise, a patch is tested by its partial velocities, a form is tested by how it eats tangent data, and a curvature formula is tested against a model where the answer is already known. That proximity is useful for learning: if a diagram changes when a parameter changes, the reader can ask which invariant stayed fixed and which part was merely a coordinate choice.

The central habit is to separate representation from geometry. Coordinates are useful because they let us compute, but the geometry usually lives in what survives a change of coordinates, a rigid motion, a reparametrization, or a different chart. The visual cells below are designed to keep that distinction visible. A plot does not certify a theorem by itself; the saved JSON checks give the plot a small audit trail.

Readers should treat the code as part of the prose. Variables use names that mirror the mathematical roles: `T` for a tangent direction, `N` for a normal, `K` for Gaussian curvature, `E`, `F`, and `G` for metric coefficients when they appear. The aim is not to hide the formulas inside a library. The aim is to make the formulas short enough that the geometric idea remains in view.
