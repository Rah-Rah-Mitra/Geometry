# J-holomorphic Curves and Symplectic Topology - Standalone Notebook Course

This course is an original executable notebook edition of McDuff and Salamon's text. The local PDF is used only for source orientation; the notebooks replace passive reading with generated diagrams, proof ledgers, numerical experiments, and invariant checks.

## Source Notes

- Source: Dusa McDuff and Dietmar Salamon, J-holomorphic Curves and Symplectic Topology, second edition.
- The local AMS PDF is used for orientation only; notebooks use original prose, generated diagrams, and executable checks.
- Printed page 1 starts on physical PDF page 16, so PDF page = printed page + 15 for main matter.
- Bibliography, symbol list, and index are inventoried as back matter rather than canonical teaching notebooks.

- [Chapter 1: Introduction](chapter-01-introduction/00-index.ipynb) - [canonical notebook](chapter-01-introduction/01-introduction.ipynb); printed pp. 1-16; PDF pp. 16-31; Symplectic manifolds, tame almost complex structures, moduli spaces, evaluation maps, and the Gromov-Witten counting pipeline.

- [Chapter 3: Moduli Spaces and Transversality](chapter-03-moduli-spaces-and-transversality/00-index.ipynb) - [canonical notebook](chapter-03-moduli-spaces-and-transversality/03-moduli-spaces-and-transversality.ipynb); printed pp. 39-74; PDF pp. 54-89; Moduli spaces of simple curves, transversality, regularity criteria, point constraints, and the implicit function theorem.
- [Chapter 4: Compactness](chapter-04-compactness/00-index.ipynb) - [canonical notebook](chapter-04-compactness/04-compactness.ipynb); printed pp. 75-114; PDF pp. 90-129; Energy, bubbling, mean value and isoperimetric inequalities, removal of singularities, convergence modulo bubbling, and connected bubble trees.
- [Chapter 5: Stable Maps](chapter-05-stable-maps/00-index.ipynb) - [canonical notebook](chapter-05-stable-maps/05-stable-maps.ipynb); printed pp. 115-152; PDF pp. 130-167; Stable maps, Gromov convergence, compactness for stable maps, uniqueness of limits, and the Gromov topology.
- [Chapter 6: Moduli Spaces of Stable Maps](chapter-06-moduli-spaces-of-stable-maps/00-index.ipynb) - [canonical notebook](chapter-06-moduli-spaces-of-stable-maps/06-moduli-spaces-of-stable-maps.ipynb); printed pp. 153-200; PDF pp. 168-215; Simple stable maps, transversality, evaluation maps, semipositivity, pseudocycles, Gromov-Witten pseudocycles, and graph pseudocycles.
- [Chapter 7: Gromov-Witten Invariants](chapter-07-gromov-witten-invariants/00-index.ipynb) - [canonical notebook](chapter-07-gromov-witten-invariants/07-gromov-witten-invariants.ipynb); printed pp. 201-256; PDF pp. 216-271; Counting pseudoholomorphic spheres, variations on the definition, graph counts, rational curves in projective spaces, and Gromov-Witten axioms.
- [Chapter 8: Hamiltonian Perturbations](chapter-08-hamiltonian-perturbations/00-index.ipynb) - [canonical notebook](chapter-08-hamiltonian-perturbations/08-hamiltonian-perturbations.ipynb); printed pp. 257-294; PDF pp. 272-309; Trivial bundles, locally Hamiltonian fibrations, pseudoholomorphic sections, fiber spheres, section pseudocycles, and section counts.
- [Chapter 9: Applications in Symplectic Topology](chapter-09-applications-in-symplectic-topology/00-index.ipynb) - [canonical notebook](chapter-09-applications-in-symplectic-topology/09-applications-in-symplectic-topology.ipynb); printed pp. 295-368; PDF pp. 310-383; Hamiltonian periodic orbits, Lagrangian embeddings, nonsqueezing, symplectic four-manifolds, symplectomorphism groups, Hofer geometry, and distinguishing symplectic structures.
- [Chapter 10: Gluing](chapter-10-gluing/00-index.ipynb) - [canonical notebook](chapter-10-gluing/10-gluing.ipynb); printed pp. 369-416; PDF pp. 384-431; Gluing theorem, connected sums of curves, weighted norms, cutoff functions, construction and derivative of the gluing map, surjectivity, splitting axiom, and a revisited theorem.
- [Chapter 11: Quantum Cohomology](chapter-11-quantum-cohomology/00-index.ipynb) - [canonical notebook](chapter-11-quantum-cohomology/11-quantum-cohomology.ipynb); printed pp. 417-486; PDF pp. 432-501; Small quantum cohomology, Gromov-Witten potential, examples, Seidel representation, and Frobenius manifolds.
- [Chapter 12: Floer Homology](chapter-12-floer-homology/00-index.ipynb) - [canonical notebook](chapter-12-floer-homology/12-floer-homology.ipynb); printed pp. 487-530; PDF pp. 502-545; Floer cochain complex, ring structure, Poincare duality, spectral invariants, Seidel representation, Donaldson's quantum category, and vortex equations.
- [Appendix A: Fredholm Theory](appendix-a-fredholm-theory/00-index.ipynb) - [canonical notebook](appendix-a-fredholm-theory/a-fredholm-theory.ipynb); printed pp. 531-548; PDF pp. 546-563; Fredholm operators, determinant line bundles, implicit function theorem, finite-dimensional reduction, and Sard-Smale theorem.
- [Appendix B: Elliptic Regularity](appendix-b-elliptic-regularity/00-index.ipynb) - [canonical notebook](appendix-b-elliptic-regularity/b-elliptic-regularity.ipynb); printed pp. 549-578; PDF pp. 564-593; Sobolev spaces, Calderon-Zygmund inequality, regularity for the Laplace operator, and elliptic bootstrapping.
- [Appendix C: The Riemann-Roch Theorem](appendix-c-riemann-roch-theorem/00-index.ipynb) - [canonical notebook](appendix-c-riemann-roch-theorem/c-riemann-roch-theorem.ipynb); printed pp. 579-618; PDF pp. 594-633; Cauchy-Riemann operators, elliptic estimates, boundary Maslov index, proof of Riemann-Roch, Riemann mapping, nonsmooth bundles, and almost complex structures.
- [Appendix D: Stable Curves of Genus Zero](appendix-d-stable-curves-of-genus-zero/00-index.ipynb) - [canonical notebook](appendix-d-stable-curves-of-genus-zero/d-stable-curves-of-genus-zero.ipynb); printed pp. 619-652; PDF pp. 634-667; Mobius transformations, cross ratios, trees, labels, splittings, stable curves, Grothendieck-Knudsen space, Gromov topology, cohomology, and examples.
- [Appendix E: Singularities and Intersections](appendix-e-singularities-and-intersections/00-index.ipynb) - [canonical notebook](appendix-e-singularities-and-intersections/e-singularities-and-intersections.ipynb); printed pp. 653-694; PDF pp. 668-709; Main intersection results, positivity of intersections, integrability, Hartman-Wintner theorem, local behavior, contact between branches, and singularities of J-holomorphic curves.
